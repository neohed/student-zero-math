# 01 — Linear Motion

At some point — in a game, a demo, a side project — you probably wrote something like this:

```python
x = 0
while running:
    x += 5
    draw_circle(x, y)
```

A variable. A loop. An increment. Something moves across the screen.

What you may not know is that you were writing a **linear equation**.

This notebook makes that connection explicit. We'll build the animation first, then reveal the math hiding inside it. By the end you'll see that `x = x0 + v * t` and the slope-intercept form $y = mx + b$ are the same equation with different names.

**Prerequisites:** Basic arithmetic. Familiarity with a loop and a variable in any programming language. No prior math required.

---

## 1. Draw Something

First, let's just get a circle on screen. Nothing moves yet.

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(10, 3))
fig.patch.set_facecolor('#1e1e2e')   # dark background
ax.set_facecolor('#1e1e2e')
ax.set_xlim(0, 100)
ax.set_ylim(0, 30)
ax.set_aspect('equal')
ax.axis('off')

# Draw a circle at position x=10, y=15
circle = plt.Circle((10, 15), radius=4, color='#89dceb')
ax.add_patch(circle)

plt.tight_layout()
plt.show()
````

---

## 2. Make It Move

Now we add the game loop. Each frame, we increase `x` by a fixed amount — the **velocity**.

Run this cell and watch what happens.

````python
# FuncAnimation requires an interactive backend — add '%matplotlib widget' above if it renders static
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# --- The game loop variables ---
x0 = 5       # starting position
v  = 2       # velocity: how much x increases each frame
t  = 0       # frame counter (time)

fig, ax = plt.subplots(figsize=(10, 3))
fig.patch.set_facecolor('#1e1e2e')
ax.set_facecolor('#1e1e2e')
ax.set_xlim(0, 100)
ax.set_ylim(0, 30)
ax.set_aspect('equal')
ax.axis('off')

circle = plt.Circle((x0, 15), radius=4, color='#89dceb')
ax.add_patch(circle)

# Frame counter label
frame_label = ax.text(2, 27, '', color='#cdd6f4', fontsize=9, fontfamily='monospace')

def update(frame):
    # This is the game loop body
    # x increases by v every frame — wraps around when it hits the edge
    x = x0 + v * frame
    x_wrapped = (x % 110) - 5   # wrap: reappear from left when off right edge
    circle.set_center((x_wrapped, 15))
    frame_label.set_text(f't (frame) = {frame:>3}   x = {x0} + {v} × {frame} = {x}')
    return circle, frame_label

ani = animation.FuncAnimation(fig, update, frames=60, interval=80, blit=True)
plt.tight_layout()
plt.show()
````

Watch the label at the top left. Each frame, `x` is calculated as:

```
x = x0 + v × t
```

That formula should look familiar. Let's name it properly.

---

## 3. The Reveal — You Were Writing This

Here is the game loop code alongside the equation it implements:

| Game Loop | Math | Role |
|-----------|------|------|
| `x = x0 + v * t` | $x = x_0 + vt$ | the equation |
| `x0` — starting position | $x_0$ — initial value | y-intercept |
| `v` — how much x changes per frame | $v$ — rate of change | slope |
| `t` — frame counter | $t$ — time | input |
| `x` — current position | $x$ — output value | output |

This is a **linear equation** — the same as the slope-intercept form $y = mx + b$ you may have seen before, just with different variable names. `v` is the slope `m`. `x0` is the y-intercept `b`.

- Double the frame count → double the distance travelled
- The rate of change is constant — it never speeds up or slows down
- Plot `x` against `t` and you get a straight line

That is what *linear* means. Constant rate of change. Straight line.

---

## 4. The Line

Let's plot the position of the circle over time. If it's linear, we should get a straight line.

````python
import matplotlib.pyplot as plt

x0 = 5
v  = 2
frames = list(range(0, 48))
positions = [x0 + v * t for t in frames]   # this is x = x0 + vt

fig, ax = plt.subplots(figsize=(10, 4))
fig.patch.set_facecolor('#1e1e2e')
ax.set_facecolor('#1e1e2e')
ax.tick_params(colors='#cdd6f4')
for spine in ax.spines.values():
    spine.set_edgecolor('#45475a')

ax.plot(frames, positions, color='#89dceb', linewidth=2)
ax.set_xlabel('t  (frame number)', color='#cdd6f4')
ax.set_ylabel('x  (position)', color='#cdd6f4')
ax.set_title('Position over time — x = x₀ + vt', color='#cdd6f4', pad=12)

plt.tight_layout()
plt.show()
````

Straight line. Constant slope. That slope *is* `v` — the velocity.

Change `v` and you change the steepness of the line. Steeper line = faster movement. Negative `v` = negative slope = movement in the opposite direction. The plot below shows all three cases at once.

````python
import matplotlib.pyplot as plt

x0     = 5
frames = list(range(0, 48))

configs = [
    {'v':  1, 'color': '#89dceb', 'label': 'v = 1  (slow)'},
    {'v':  2, 'color': '#a6e3a1', 'label': 'v = 2  (medium)'},
    {'v':  4, 'color': '#fab387', 'label': 'v = 4  (fast)'},
    {'v': -1, 'color': '#f38ba8', 'label': 'v = −1 (moving left)'},
]

fig, ax = plt.subplots(figsize=(10, 4))
fig.patch.set_facecolor('#1e1e2e')
ax.set_facecolor('#1e1e2e')
ax.tick_params(colors='#cdd6f4')
for spine in ax.spines.values():
    spine.set_edgecolor('#45475a')

for cfg in configs:
    positions = [x0 + cfg['v'] * t for t in frames]
    ax.plot(frames, positions, color=cfg['color'], linewidth=2, label=cfg['label'])

ax.axhline(0, color='#45475a', linewidth=0.8, linestyle='--')
ax.set_xlabel('t  (frame number)', color='#cdd6f4')
ax.set_ylabel('x  (position)', color='#cdd6f4')
ax.set_title('Different velocities → different slopes', color='#cdd6f4', pad=12)
ax.legend(facecolor='#313244', labelcolor='#cdd6f4', framealpha=0.8)

plt.tight_layout()
plt.show()
````

## 5. Worked Examples

The formula is $x = x_0 + vt$. Let's use it directly with three different sets of values and check that the numbers match what the animation produced.

**Example 1** — Start at 5, velocity 2, at frame 10:

$$x = 5 + 2 \times 10 = 5 + 20 = 25$$

**Example 2** — Start at 0, velocity 3, at frame 10:

$$x = 0 + 3 \times 10 = 0 + 30 = 30$$

**Example 3** — Start at 20, velocity 1, at frame 10:

$$x = 20 + 1 \times 10 = 20 + 10 = 30$$

Notice that Examples 2 and 3 both produce `x = 30` at frame 10 — different starting positions and velocities, same result at the same moment. This is the kind of thing equations let you reason about without running the code.

````python
# Verify the worked examples in code
# Formula: x = x0 + v * t

def position(x0, v, t):
    return x0 + v * t

print(f"Example 1 — x0=5,  v=2, t=10 → x = {position(5,  2, 10)}")
print(f"Example 2 — x0=0,  v=3, t=10 → x = {position(0,  3, 10)}")
print(f"Example 3 — x0=20, v=1, t=10 → x = {position(20, 1, 10)}")
````

The code and the equation produce identical results. They are the same thing.

The equation is just a more compact notation for what the function does.

## Formal Notation

The equation you have been writing:

$$x = x_0 + vt$$

is a **linear equation** in $t$. It has the same form as slope-intercept:

$$y = mx + b$$

with different variable names for the same roles:

| This notebook | Slope-intercept | Meaning |
|---------------|----------------|---------|
| $x$ | $y$ | output (position) |
| $t$ | $x$ | input (time / frame count) |
| $v$ | $m$ | **slope** — rate of change |
| $x_0$ | $b$ | **y-intercept** — starting value |

The slope $m$ (or $v$) is the one number that defines the line's steepness. It is the answer to: *how much does the output change for every one-unit increase in the input?* Constant slope is exactly what makes an equation linear — and exactly what makes a fixed-increment loop draw a straight line.

---

## Real-World Connection

- **Physics engines:** every rigid body has a velocity vector. Each timestep, position is updated as `pos += velocity * dt` — this exact formula, in 2D or 3D with a real time delta instead of a frame counter.
- **GPS dead-reckoning:** when satellite signal is briefly lost, a phone estimates position using its last known speed and heading. The estimate is `x = x₀ + v × elapsed_time`.
- **CSS animations:** `animation-timing-function: linear` means the animated property advances at a constant rate per millisecond. The browser computes `x = x₀ + v × t` every frame.
- **`lerp` in graphics:** the `lerp(a, b, t)` function used everywhere in rendering is `a + (b - a) * t` — which is `a + v * t` where `v = b - a`. Linear motion and linear interpolation are the same equation.
- **Normalisation and scaling:** any time you map one range of values onto another at a constant rate, you are computing a linear equation.

---

## Summary

| Concept | Plain English | Code | Math |
|---------|--------------|------|------|
| Linear equation | Output changes at a constant rate | `x = x0 + v * t` | $x = x_0 + vt$ |
| Velocity / slope | How much position changes per frame | `v` | $v$ (same as $m$ in $y = mx + b$) |
| Initial position / intercept | Where you start | `x0` | $x_0$ (same as $b$ in $y = mx + b$) |
| Linear | Straight line, constant rate of change | increment never changes | slope is constant |

**The key insight:** a loop variable that increments by a constant amount *is* a linear equation. You have been writing linear equations every time you moved something across a screen.
