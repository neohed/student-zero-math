# Order of Operations

Without agreed-upon rules for evaluation order, the expression `2 + 3 * 4` is ambiguous. It could mean `(2 + 3) * 4 = 20` or `2 + (3 * 4) = 14`. Those are different answers.

Order of operations is the set of rules that resolves this ambiguity. Every language that has arithmetic — Python, C, SQL, Excel — uses the same rules, inherited from centuries of mathematical convention.

You already rely on these rules every time you write an expression in code. This notebook makes them explicit.

**Prerequisites:** Basic Python. Basic arithmetic.

---

## The Core Problem: Ambiguity

Imagine you are writing a parser for a simple calculator. Your input is the string `"2 + 3 * 4"`. What does it evaluate to?

There are two equally valid-looking interpretations:

```
Left-to-right:   (2 + 3) * 4 = 5 * 4 = 20
Multiply first:   2 + (3 * 4) = 2 + 12 = 14
```

Without a rule to break the tie, the same expression means different things to different people — or different programs.

The solution is **operator precedence**: a ranking that tells you which operations to perform before others. It's exactly the same concept as operator precedence in a programming language. Python has a documented precedence table. So does mathematics. They agree.

````python
# Python evaluates this the same way mathematics does
print(2 + 3 * 4)     # 14 — multiplication happens first
print((2 + 3) * 4)   # 20 — parentheses force addition first

# Python's precedence is not magic — it follows the same rules you're about to learn
````

---

## The Rules: PEMDAS

The conventional order is remembered with the acronym **PEMDAS** (also taught as BODMAS in the UK):

| Priority | Name | Symbols | Example |
|----------|------|---------|---------|
| 1 (highest) | **P**arentheses | `( )` | `(2 + 3)` evaluated first |
| 2 | **E**xponents | `**`, `^` | `2 ** 3` = 8 |
| 3 | **M**ultiplication and **D**ivision | `*`, `/`, `//`, `%` | Left to right |
| 4 (lowest) | **A**ddition and **S**ubtraction | `+`, `-` | Left to right |

Two things to notice:

1. **M and D are the same level** — multiplication and division don't have priority over each other; ties are broken left-to-right.
2. **A and S are the same level** — same rule applies to addition and subtraction.

Think of it as a four-tier system, not six steps.

---

## Parentheses: Override Everything

Parentheses have the highest priority. Whatever is inside them is evaluated first, regardless of what operators surround them.

Think of parentheses as function call grouping — in Python, `f(x + y)` computes `x + y` before passing it to `f`. Same idea: the inner expression resolves before the outer one can proceed.

Parentheses can be nested. Innermost wins first.

````python
# Parentheses override default precedence
print(2 + 3 * 4)        # 14 — default: multiply first
print((2 + 3) * 4)      # 20 — parens force addition first

print()

# Nested parentheses — innermost resolves first
expr = (2 + (3 * (4 + 1)))
# Step 1: (4 + 1) = 5
# Step 2: (3 * 5) = 15
# Step 3: (2 + 15) = 17
print(expr)              # 17

# Let's trace it step by step
step1 = 4 + 1
step2 = 3 * step1
step3 = 2 + step2
print(f"Step 1: 4 + 1 = {step1}")
print(f"Step 2: 3 * {step1} = {step2}")
print(f"Step 3: 2 + {step2} = {step3}")
````

---

## Exponents: Right to Left

Exponents are evaluated before multiplication and addition, but after parentheses.

There is one subtlety: when you chain exponents, they evaluate **right to left**. This is called right-associativity, and it matches the mathematical convention:

```
2 ** 3 ** 2   means   2 ** (3 ** 2)   =   2 ** 9   =   512
```

Not:

```
(2 ** 3) ** 2   =   8 ** 2   =   64
```

This is the only operator in the standard arithmetic set that is right-associative. Everything else (multiplication, addition, etc.) goes left to right.

````python
# Exponents bind tighter than * and +
print(2 + 3 ** 2)        # 11 — exponent first: 3**2=9, then 2+9
print((2 + 3) ** 2)      # 25 — parens first: (2+3)=5, then 5**2
print(2 * 3 ** 2)        # 18 — exponent first: 3**2=9, then 2*9

print()

# Right-associativity of exponents
a = 2 ** 3 ** 2          # right-to-left: 2 ** (3 ** 2) = 2 ** 9
b = (2 ** 3) ** 2        # forced left: 8 ** 2

print(f"2 ** 3 ** 2       = {a}   (right-associative: 2 ** 9)")
print(f"(2 ** 3) ** 2     = {b}   (forced left: 8 ** 2)")
print(f"Are they equal?   {a == b}")
````

---

## Multiplication and Division: Same Level, Left to Right

Multiplication and division are at equal priority. When they appear together in an expression, you evaluate **left to right** — exactly as you'd read the expression.

```
12 / 4 * 3   →   (12 / 4) * 3   =   3 * 3   =   9
```

Not `12 / (4 * 3) = 1`. Left to right.

This catches people out, especially when writing formulas. `a / b * c` and `a / (b * c)` give very different answers.

````python
# Multiplication and division: left to right
print(12 / 4 * 3)        # 9.0 — (12/4)*3, not 12/(4*3)
print(12 / (4 * 3))      # 1.0 — forced grouping

print()

# Integer division and modulo follow the same rule
print(20 // 3 * 2)       # (20 // 3) * 2 = 6 * 2 = 12
print(10 % 3 * 2)        # (10 % 3) * 2 = 1 * 2 = 2

print()

# A common bug: writing a fraction formula without parentheses
a, b, c = 10, 4, 2

wrong   = a / b + c      # looks like a/(b+c) but isn't — evaluates as (a/b) + c = 4.5
correct = a / (b + c)    # 10 / 6 ≈ 1.667

print(f"a / b + c   = {wrong}     <- NOT the fraction a/(b+c)")
print(f"a / (b + c) = {correct:.4f}  <- correct fraction")
````

---

## Addition and Subtraction: Same Level, Left to Right

Addition and subtraction are the lowest-priority operations, and again they are equal to each other. Left to right when they appear together.

```
10 - 3 + 2   →   (10 - 3) + 2   =   7 + 2   =   9
```

Not `10 - (3 + 2) = 5`.

Subtraction is the operator that tricks people. It feels like it should bind its right operand tightly, but it doesn't — it's just left-to-right, same as everything else at this level.

````python
# Addition and subtraction: left to right
print(10 - 3 + 2)        # 9 — (10-3)+2, not 10-(3+2)
print(10 - (3 + 2))      # 5 — forced grouping

print()

# Unary minus (negation) is its own operation, applied to a single operand
# It has higher precedence than binary addition/subtraction
print(-3 + 5)            # 2 — (-3) + 5
print(-(3 + 5))          # -8 — negate the whole group

print()

# Step through a mixed expression to build intuition
expr_value = 8 - 2 + 1 - 3
#            (8 - 2) = 6
#            (6 + 1) = 7
#            (7 - 3) = 4
print(f"8 - 2 + 1 - 3 = {expr_value}")
print(f"Step by step: ({8} - {2}) = {8-2}, ({8-2} + {1}) = {8-2+1}, ({8-2+1} - {3}) = {8-2+1-3}")
````

---

## Putting It All Together

Let's evaluate a complex expression step by step, applying the rules in order:

```
3 + 2 ** 3 * (10 - 4) / 6 - 1
```

Step 1 — **Parentheses**: `(10 - 4)` = `6`
```
3 + 2 ** 3 * 6 / 6 - 1
```

Step 2 — **Exponents**: `2 ** 3` = `8`
```
3 + 8 * 6 / 6 - 1
```

Step 3 — **Multiplication/Division** (left to right): `8 * 6` = `48`, then `48 / 6` = `8`
```
3 + 8 - 1
```

Step 4 — **Addition/Subtraction** (left to right): `3 + 8` = `11`, then `11 - 1` = `10`
```
10
```

````python
# Verify the worked example
result = 3 + 2 ** 3 * (10 - 4) / 6 - 1
print(f"3 + 2 ** 3 * (10 - 4) / 6 - 1 = {result}")

print()

# Trace it step by step using intermediate variables
step1 = 10 - 4         # parentheses
step2 = 2 ** 3         # exponent
step3 = step2 * step1  # left-to-right: multiply
step4 = step3 / 6      # left-to-right: divide
step5 = 3 + step4      # left-to-right: add
step6 = step5 - 1      # left-to-right: subtract

print(f"Step 1 (parens):     10 - 4          = {step1}")
print(f"Step 2 (exponent):   2 ** 3          = {step2}")
print(f"Step 3 (multiply):   {step2} * {step1}           = {step3}")
print(f"Step 4 (divide):     {step3} / 6          = {step4}")
print(f"Step 5 (add):        3 + {step4}         = {step5}")
print(f"Step 6 (subtract):   {step5} - 1           = {step6}")
````

## Visualisation: The Evaluation Tree

A good way to see operator precedence is as an **expression tree** (also called an abstract syntax tree — the same structure a compiler builds when parsing your code).

Operators with lower precedence sit higher in the tree; operators with higher precedence sit lower (closer to the leaves). Evaluation always starts at the leaves and works upward.

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Draw a simple expression tree for: 2 + 3 * 4
# Tree structure:
#        +
#       / \
#      2   *
#         / \
#        3   4

fig, ax = plt.subplots(figsize=(7, 5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis('off')

def draw_node(ax, x, y, label, color='steelblue', text_color='white'):
    circle = plt.Circle((x, y), 0.5, color=color, zorder=3)
    ax.add_patch(circle)
    ax.text(x, y, label, ha='center', va='center',
            fontsize=14, fontweight='bold', color=text_color, zorder=4)

def draw_edge(ax, x1, y1, x2, y2):
    ax.plot([x1, x2], [y1, y2], 'k-', linewidth=1.5, zorder=2)

# Nodes: (x, y, label, color)
nodes = [
    (5, 5,   '+',  'steelblue'),    # root: addition (lowest precedence)
    (3, 3.2, '2',  '#888888'),      # left leaf
    (7, 3.2, '*',  'darkorange'),   # right child: multiplication (higher)
    (6, 1.5, '3',  '#888888'),      # leaf
    (8, 1.5, '4',  '#888888'),      # leaf
]

# Edges
edges = [
    (5, 5, 3, 3.2),
    (5, 5, 7, 3.2),
    (7, 3.2, 6, 1.5),
    (7, 3.2, 8, 1.5),
]

for x1, y1, x2, y2 in edges:
    draw_edge(ax, x1, y1, x2, y2)

for x, y, label, color in nodes:
    draw_node(ax, x, y, label, color)

# Annotations
ax.text(5, 5.7,  'evaluates last (lowest precedence)',
        ha='center', fontsize=9, color='steelblue')
ax.text(7, 2.45, 'evaluates first (higher precedence)',
        ha='center', fontsize=9, color='darkorange')
ax.text(5, 0.5, 'Expression: 2 + 3 * 4 = 14',
        ha='center', fontsize=12, style='italic')

ax.set_title('Expression tree: operators with higher precedence sit deeper in the tree',
             fontsize=11)
plt.tight_layout()
plt.show()
````

````python
# Python's `ast` module lets you inspect the actual expression tree
# that Python builds when it parses your code.
import ast

def show_tree(expression_str):
    tree = ast.parse(expression_str, mode='eval')
    print(f"Expression: {expression_str}")
    print(ast.dump(tree, indent=2))
    print()

show_tree("2 + 3 * 4")
show_tree("(2 + 3) * 4")
````

The `ast.dump` output shows that Python parses `2 + 3 * 4` with the multiplication as a sub-node of the addition — which means Python evaluates the multiplication first, exactly as the rules dictate. Parentheses in `(2 + 3) * 4` flip the tree: now the addition is the deeper node.

## Formal Notation

Mathematics uses implicit ordering conventions rather than named rules, but the priority is the same:

$$3 + 2^3 \times (10 - 4) \div 6 - 1$$

Evaluated as:

$$3 + 2^3 \times 6 \div 6 - 1 = 3 + 8 \times 6 \div 6 - 1 = 3 + 8 - 1 = 10$$

Formal priority order (highest to lowest):

$$\text{(grouping)} > \text{exponentiation} > \{\times, \div\} > \{+, -\}$$

Where $\{\times, \div\}$ and $\{+, -\}$ are evaluated **left to right** within their level.

Exponentiation chains evaluate **right to left**: $a^{b^c} = a^{(b^c)}$

---

## Real-World Connection

Order of operations is not abstract. It determines whether your code is correct:

- **Formula bugs**: The expression for the average of two integers, `(a + b) / 2`, is often written as `a + b / 2` by mistake — a different formula entirely. The same bug appears in midpoint calculations, interpolation, and statistics.
- **Parsing and compilers**: Every language parser contains a precedence table. When Python reads your source code, it builds an AST respecting exactly these rules. Writing a parser (for a config file, a DSL, or a query language) requires you to encode precedence explicitly.
- **Spreadsheets**: Excel and Google Sheets use the same precedence rules. `=A1+B1*C1` is not the same as `=(A1+B1)*C1`. Spreadsheet bugs due to missing parentheses are common and sometimes expensive.
- **Distributivity from Notebook 01** relies on order of operations being defined: `a * (b + c) = a*b + a*c` only makes sense because we have agreed rules for which operation binds first.
- **Operator overloading**: When you define `__add__` and `__mul__` in Python, the language still applies its precedence table to decide which is called first. Your custom objects inherit the standard ordering.

---

## Summary

- Without agreed evaluation order, arithmetic expressions are ambiguous. Order of operations is the convention that eliminates the ambiguity.
- **PEMDAS** gives the priority: **P**arentheses > **E**xponents > **M**ultiplication and **D**ivision > **A**ddition and **S**ubtraction.
- Multiplication and Division are the same level — tie broken left to right.
- Addition and Subtraction are the same level — tie broken left to right.
- Exponents chain **right to left**: `a ** b ** c` = `a ** (b ** c)`.
- Python's operator precedence table implements exactly these rules. Parentheses are how you override them.
- Every compiler and parser encodes these rules as a precedence table. Understanding them makes you a better reader and writer of expressions — in any language.
