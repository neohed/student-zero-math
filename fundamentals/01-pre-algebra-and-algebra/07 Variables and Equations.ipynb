# Variables and Equations

A variable is a placeholder for an unknown value. An equation is an assertion that two expressions are equal. Solving an equation means finding the value that makes the assertion true.

You have been using variables in this exact sense since the first day you wrote code. The only new thing here is the algebraic manipulation technique — a single rule that lets you isolate any unknown in any equation.

This notebook also makes explicit what several earlier notebooks (03, 06) and all the probability notebooks have been doing silently: **rearranging formulas**. Deriving the reciprocal rule, cross multiplication, and Bayes' theorem all use the same move. This notebook names it.

**Prerequisites:** Notebooks 00–06.

## Variables: Placeholders for Unknown Values

In Python, a function parameter is a placeholder — it has no value until someone calls the function:

```python
def f(x):
    return 2 * x + 3
```

The `x` here is a mathematical variable. It does not have a value yet. The function describes a *relationship* between input and output, not a specific number.

In mathematics, variables appear in exactly the same role: as names for values we either do not know yet or want to reason about generally. The letter is arbitrary — `x`, `n`, `t`, `price` — convention just tends toward single letters.

An **expression** contains variables and operations but no assertion:

```python
2 * x + 3       # an expression — has a value, but we haven't fixed x
```

An **equation** asserts that two expressions are equal:

```python
2 * x + 3 == 11  # an equation — an assertion that may or may not be true
```

Solving the equation means finding the value of `x` that makes the assertion `True`.

````python
# An expression: computable for any x, but x must be given a value first
def expr(x):
    return 2 * x + 3

print("Expression 2x + 3 for various x:")
for x in range(-2, 6):
    print(f"  x = {x:2d}  =>  2({x}) + 3 = {expr(x)}")
````

````python
# An equation: which x makes 2x + 3 == 11 true?
print("Which x satisfies 2x + 3 = 11?")
for x in range(-2, 10):
    result = 2 * x + 3
    satisfies = result == 11
    marker = "  <-- solution" if satisfies else ""
    print(f"  x = {x:2d}  =>  2({x}) + 3 = {result:3d}  ==  11? {satisfies}{marker}")
````

Brute-force search found `x = 4`. Algebra finds it directly, in one step.

## The Only Rule: Do the Same Thing to Both Sides

An equation is a balance. Both sides are equal. If you apply the same operation to both sides, the balance is preserved.

In code: if `a == b`, then for any function `f`, `f(a) == f(b)`.

That's it. That is the entire foundation of algebraic manipulation. Every technique — isolating a variable, rearranging a formula, deriving a new rule from an existing one — is just this rule applied repeatedly.

The goal is always the same: apply operations to both sides until the unknown is alone on one side.

```
2x + 3 = 11

Subtract 3 from both sides:    2x + 3 - 3 = 11 - 3
                                       2x = 8

Divide both sides by 2:             2x/2 = 8/2
                                        x = 4
```

The unknown is now isolated. Each step preserved the equality.

````python
from fractions import Fraction

def solve_linear_trace(a, b, c):
    """
    Solve ax + b = c, printing each step.
    Applies the balance rule twice:
      1. Subtract b from both sides
      2. Divide both sides by a
    """
    a, b, c = Fraction(a), Fraction(b), Fraction(c)

    print(f"  Equation:              {a}x + {b} = {c}")

    # Step 1: subtract b from both sides
    rhs_after_sub = c - b
    print(f"  Subtract {b} both sides: {a}x + {b} - {b} = {c} - {b}")
    print(f"                           {a}x = {rhs_after_sub}")

    # Step 2: divide both sides by a
    x = rhs_after_sub / a
    print(f"  Divide by {a} both sides: {a}x / {a} = {rhs_after_sub} / {a}")
    print(f"                           x = {x}")

    # Verify
    check = a * x + b
    print(f"  Verify: {a}({x}) + {b} = {check}  ==  {c}? {check == c}")
    print()
    return x

solve_linear_trace(2, 3, 11)    # 2x + 3 = 11
solve_linear_trace(3, -5, 10)   # 3x - 5 = 10
solve_linear_trace(1, 7, 2)     # x + 7 = 2
````

## The Order of Unwrapping

Notice the order: to solve `2x + 3 = 11`, we subtracted first, then divided. The operations were applied in **reverse order** compared to how the expression was built.

The expression `2x + 3` applies multiplication first (multiply x by 2), then addition (add 3). Unwrapping reverses this: undo the addition first, then undo the multiplication.

This is exactly like unwrapping nested function calls:

```python
# Forward:  y = add3(times2(x))      =>  y = 2x + 3
# Unwrap:   x = inv_times2(inv_add3(y))
#           x = (y - 3) / 2
```

Undo the outermost operation first, work inward. The order of operations (notebook 00) determines the order of unwrapping — they go in opposite directions.

````python
# Forward and inverse functions — the algebraic unwrapping is the function inverse

def forward(x):
    """Build the expression: 2x + 3"""
    step1 = 2 * x       # times 2 first
    step2 = step1 + 3   # add 3 second
    return step2

def inverse(y):
    """Undo the expression: solve for x given y = 2x + 3"""
    step1 = y - 3       # undo addition FIRST (outermost)
    step2 = step1 / 2   # undo multiplication SECOND (innermost)
    return step2

print("Forward and inverse are symmetric:")
for x in [0, 1, 4, -3, 2.5]:
    y = forward(x)
    x_back = inverse(y)
    print(f"  x = {x:5}  ->  f(x) = {y:5}  ->  f⁻¹(y) = {x_back}")
````

## Variables on Both Sides

When the unknown appears on both sides, the first move is to collect all the variable terms on one side:

```
3x + 2 = x + 8

Subtract x from both sides:    3x - x + 2 = x - x + 8
                                    2x + 2 = 8

Subtract 2 from both sides:      2x + 2 - 2 = 8 - 2
                                          2x = 6

Divide both sides by 2:                    x = 3
```

The same balance rule, applied three times. No new rule required.

````python
from fractions import Fraction

def solve_both_sides_trace(a1, b1, a2, b2):
    """
    Solve  a1*x + b1 = a2*x + b2
    Collect x terms on the left, constants on the right.
    """
    a1, b1, a2, b2 = Fraction(a1), Fraction(b1), Fraction(a2), Fraction(b2)

    print(f"  Equation:               {a1}x + {b1} = {a2}x + {b2}")

    # Step 1: subtract a2*x from both sides
    a_new = a1 - a2
    print(f"  Subtract {a2}x both sides:  {a_new}x + {b1} = {b2}")

    # Step 2: subtract b1 from both sides
    b_new = b2 - b1
    print(f"  Subtract {b1} both sides:   {a_new}x = {b_new}")

    # Step 3: divide both sides by a_new
    x = b_new / a_new
    print(f"  Divide by {a_new} both sides: x = {x}")

    check_left  = a1 * x + b1
    check_right = a2 * x + b2
    print(f"  Verify: {a1}({x}) + {b1} = {check_left}  ==  {a2}({x}) + {b2} = {check_right}? {check_left == check_right}")
    print()
    return x

solve_both_sides_trace(3, 2, 1, 8)    # 3x + 2 = x + 8
solve_both_sides_trace(5, -1, 2, 5)   # 5x - 1 = 2x + 5
````

## Visualisation: An Equation is an Intersection

Solving `2x + 3 = 11` can be read as: where does the line `y = 2x + 3` cross the horizontal line `y = 11`?

The x-coordinate of that intersection is the solution. This is why algebra and geometry are the same thing — an equation describes a constraint, and its solution is the point that satisfies it.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 6, 200)
y_expr = 2 * x + 3     # left-hand side
y_rhs  = 11            # right-hand side

solution_x = (11 - 3) / 2   # = 4
solution_y = 2 * solution_x + 3

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x, y_expr, 'steelblue', linewidth=2, label='y = 2x + 3  (left side)')
ax.axhline(y_rhs, color='darkorange', linewidth=2, linestyle='--', label=f'y = {y_rhs}  (right side)')

ax.plot(solution_x, solution_y, 'ro', markersize=10, zorder=5, label=f'Solution: x = {solution_x}')
ax.axvline(solution_x, color='red', linewidth=1, linestyle=':', alpha=0.6)
ax.text(solution_x + 0.1, 2, f'x = {solution_x}', color='red', fontsize=11)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Solving 2x + 3 = 11: find where the two lines cross', fontsize=11)
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## Rearranging Formulas

The balance rule applies to formulas, not just simple equations. A formula is an equation with multiple variables — and you can solve for any one of them.

This is what earlier notebooks have been doing silently:

- **Notebook 03**: Starting from `(c/d) × x = a/b`, multiply both sides by `d/c` to isolate `x`. That's the reciprocal rule derived.
- **Notebook 06**: Starting from `a/b = c/d`, multiply both sides by `bd` to get `ad = bc`. That's cross multiplication derived.
- **Notebook 11**: Starting from `P(A|B) = P(A∩B)/P(B)`, rearranging twice gives Bayes' theorem.

These were not tricks. They were all the same rule: do the same thing to both sides, choosing the operation that moves you toward the unknown being isolated.

Here is the slope-intercept formula for a line, solved for each of its variables in turn:

````python
# y = mx + b
# Rearrange for each variable — same balance rule each time.

# Solve for x:
#   y = mx + b
#   y - b = mx          (subtract b from both sides)
#   (y - b) / m = x     (divide both sides by m)

# Solve for m:
#   y = mx + b
#   y - b = mx          (subtract b from both sides)
#   (y - b) / x = m     (divide both sides by x)

# Solve for b:
#   y = mx + b
#   y - mx = b          (subtract mx from both sides)

from fractions import Fraction

m, b = Fraction(3, 2), Fraction(2)

# Pick a point on the line to verify all rearrangements
x_val = Fraction(4)
y_val = m * x_val + b    # = 3/2 * 4 + 2 = 8

print(f"Line: y = {m}x + {b}")
print(f"Point on line: x = {x_val}, y = {y_val}")
print()

x_solved = (y_val - b) / m
m_solved = (y_val - b) / x_val
b_solved = y_val - m * x_val

print(f"Solve for x: x = (y - b) / m  =  ({y_val} - {b}) / {m}  =  {x_solved}")
print(f"Solve for m: m = (y - b) / x  =  ({y_val} - {b}) / {x_val}  =  {m_solved}")
print(f"Solve for b: b = y - mx       =  {y_val} - {m}×{x_val}  =  {b_solved}")
````

Every formula in mathematics can be rearranged this way. The formula is not a fixed object — it is a relationship between variables that you can express from any variable's point of view.

## Formal Notation

A **linear equation in one variable** has the form:

$$ax + b = c \quad \text{where } a \neq 0$$

Solution:

$$ax + b = c$$
$$ax = c - b \qquad \text{(subtract } b \text{ from both sides)}$$
$$x = \frac{c - b}{a} \qquad \text{(divide both sides by } a \text{)}$$

The **balance axiom**: for any equation $A = B$ and any operation $f$:

$$A = B \implies f(A) = f(B)$$

provided $f$ is defined for both sides. Useful operations: $+k$, $-k$, $\times k$ (for $k \neq 0$), $\div k$ (for $k \neq 0$).

An equation with variables on both sides:

$$a_1 x + b_1 = a_2 x + b_2$$
$$(a_1 - a_2)x = b_2 - b_1$$
$$x = \frac{b_2 - b_1}{a_1 - a_2} \quad \text{for } a_1 \neq a_2$$

## Real-World Connection

- **Solving for unknowns in formulas**: every physics engine, graphics renderer, and financial model contains formulas with multiple variables. Given all but one, you solve for the last. `F = ma` → solve for `a`: `a = F/m`. Same rule.
- **Linear regression**: fitting `y = mx + b` to data means solving for `m` and `b` — rearranging the formula to find the values that minimise error. The algebra here is a direct extension of solving a two-variable equation.
- **Configuration and constraint systems**: UI layout engines (CSS, auto-layout) set up systems of linear equations where element positions and sizes are variables, and constraints are equations. Solving means finding values satisfying all constraints simultaneously.
- **Unit conversion**: `miles = km / 1.609` rearranged is `km = miles × 1.609`. Every conversion formula is a one-variable equation solved for a different unknown.
- **Inverting transformations**: game engines constantly invert transformations — going from world coordinates to screen coordinates and back. Inversion is solving an equation for the input given the output.

## Summary

- A **variable** is a placeholder for an unknown value — the same concept as a function parameter in code.
- An **expression** has a value but no assertion. An **equation** asserts two expressions are equal.
- **Solving** means finding the value that makes the assertion true — the input that passes the test.
- The **only rule**: do the same thing to both sides. This preserves equality and lets you isolate any variable.
- **Unwrap in reverse order**: operations are undone in the opposite order they were applied — outermost first.
- **Any formula can be rearranged**: the balance rule applies to formulas with multiple variables as much as to simple equations. Choosing which variable to solve for is a matter of context.
