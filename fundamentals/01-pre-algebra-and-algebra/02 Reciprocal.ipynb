# The Reciprocal

The reciprocal of a number is its **multiplicative inverse** — the value you multiply it by to get 1.

That's it. But understanding *why* that definition is useful, and where it comes from, makes a lot of subsequent algebra click into place — especially when fractions and division show up.

**Prerequisites:** Notebook 01 (Arithmetic Properties). Basic fractions.

---

## Intuition: Two Kinds of Inverse

In programming, "inverse" means undoing an operation. There are two natural inverses in arithmetic, one for each of the two fundamental operations:

| Operation | Identity | Inverse of `x` | Why |
|-----------|----------|-----------------|-----|
| Addition  | `0` | `-x` | `x + (-x) == 0` |
| Multiplication | `1` | `1/x` | `x * (1/x) == 1` |

The **additive inverse** of `x` is `-x`. Add them together, you get the additive identity (zero).

The **multiplicative inverse** of `x` is `1/x`. Multiply them together, you get the multiplicative identity (one).

The reciprocal is just the multiplicative inverse. It is the number that *undoes* multiplication — the same way `-x` undoes addition.

Notice the parallel: `0` plays the same role for addition that `1` plays for multiplication. Both are identity elements — operating with them leaves the other value unchanged (`x + 0 == x`, `x * 1 == x`).

```python
# The two inverses, side by side

x = 8

additive_inverse       = -x
multiplicative_inverse = 1 / x

print(f"x = {x}")
print(f"Additive inverse:       {additive_inverse}    =>  {x} + ({additive_inverse}) = {x + additive_inverse}")
print(f"Multiplicative inverse: {multiplicative_inverse}  =>  {x} * {multiplicative_inverse} = {x * multiplicative_inverse}")
```

---

## Why x * (1/x) = 1

Any number can be written as itself over 1 — dividing by 1 changes nothing:

$$\frac{N}{1} = N$$

So when you multiply a number by its reciprocal, you are multiplying two fractions:

$$\frac{8}{1} \times \frac{1}{8} = \frac{8 \times 1}{1 \times 8} = \frac{8}{8} = 1$$

The numerator and denominator become identical — and any non-zero number divided by itself is 1.

```python
def reciprocal(x):
    if x == 0:
        return None  # reciprocal of 0 is undefined
    return 1 / x

test_values = [8, -17, 0.5, 100, -3, 0]

print(f"{'x':>8}  {'1/x':>10}  {'x * (1/x)':>12}")
print("-" * 36)
for x in test_values:
    r = reciprocal(x)
    product = x * r if r is not None else "undefined"
    print(f"{x:>8}  {str(r):>10}  {str(product):>12}")
```

A few things to notice:

- The reciprocal of a **large number** is a small number, and vice versa
- The reciprocal of a **negative number** is also negative — the sign is preserved
- The reciprocal of a **fraction** flips it: `1 / (1/2) = 2`
- `0` has no reciprocal — there is no number you can multiply by zero to get 1

## Visualisation: The Reciprocal Curve

Plot `y = 1/x` and a pattern jumps out: the curve is symmetric around the line `y = x`. Every point `(a, b)` on the curve has a mirror point `(b, a)` — because if `b = 1/a`, then `a = 1/b`.

This symmetry is the reciprocal relationship made visible: `a` and `b` are each other's reciprocal.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.15, 5, 400)
y = 1 / x

fig, ax = plt.subplots(figsize=(7, 6))

ax.plot(x, y, 'steelblue', linewidth=2, label='y = 1/x')
ax.plot(x, x, color='lightgray', linewidth=1, linestyle='--', label='y = x (symmetry line)')

# Annotate a few reciprocal pairs
pairs = [(0.5, 2.0), (1.0, 1.0), (2.0, 0.5), (4.0, 0.25)]
for a, b in pairs:
    ax.plot(a, b, 'o', color='darkorange', markersize=7)
    ax.annotate(f'({a}, {b})', xy=(a, b), xytext=(a + 0.15, b + 0.15), fontsize=9)

ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.set_xlabel('x')
ax.set_ylabel('1/x')
ax.set_title('y = 1/x: reciprocal pairs mirror around y = x')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## The Reciprocal of a Fraction

When the number is already a fraction, taking its reciprocal simply **flips it** — numerator and denominator swap:

$$\text{reciprocal of } \frac{a}{b} = \frac{b}{a}$$

Check that it works:

$$\frac{a}{b} \times \frac{b}{a} = \frac{a \times b}{b \times a} = \frac{ab}{ab} = 1$$

This is worth holding on to — it is the foundation of the next notebook (dividing by a fraction).

```python
from fractions import Fraction

fracs = [Fraction(2, 3), Fraction(5, 4), Fraction(1, 7), Fraction(3, 1)]

print(f"{'Fraction':>10}  {'Reciprocal':>12}  {'Product':>10}")
print("-" * 38)
for f in fracs:
    r = Fraction(f.denominator, f.numerator)
    print(f"{str(f):>10}  {str(r):>12}  {str(f * r):>10}")
```

## Formal Notation

The reciprocal of $x$ (for $x \neq 0$) is written as $\frac{1}{x}$ or $x^{-1}$.

The defining property:

$$x \times \frac{1}{x} = 1 \quad \text{for } x \neq 0$$

For a fraction $\frac{a}{b}$:

$$\left(\frac{a}{b}\right)^{-1} = \frac{b}{a} \quad \text{for } a, b \neq 0$$

The notation $x^{-1}$ generalises beyond numbers — you will see it used for inverse functions ($f^{-1}$) and inverse matrices ($A^{-1}$), always meaning "the thing that undoes this operation."

---

## Real-World Connection

- **GPU shaders and 3D graphics**: division is expensive in hardware. A common optimisation is to precompute the reciprocal and multiply instead — `x * (1/y)` is faster than `x / y` on many architectures. The famous [fast inverse square root](https://en.wikipedia.org/wiki/Fast_inverse_square_root) from Quake III Arena is a hand-optimised approximation of `1 / sqrt(x)` — a reciprocal — computed using a bit-level trick.
- **Unit conversion**: multiplying by a reciprocal is often more readable than dividing. Converting km to miles? Multiply by `1/1.60934`. Conversion factors are frequently written as fractions precisely because flipping them gives the inverse conversion.
- **Matrix inversion**: the inverse of a matrix $A$ is written $A^{-1}$ and satisfies $A \cdot A^{-1} = I$ (the identity matrix). This is the exact same concept at a larger scale — finding the multiplicative inverse in a world where the "numbers" are matrices.
- **Division as multiplication**: under the hood, `a / b` is `a * (1/b)`. Every division is a multiplication by the reciprocal. That equivalence is what the next notebook builds on.

## Summary

- The reciprocal of `x` is `1/x` — the number that multiplies with `x` to give 1
- It is the multiplicative inverse, parallel to `-x` being the additive inverse
- For a fraction `a/b`, the reciprocal is `b/a` — just flip it
- Zero has no reciprocal (there is no number you can multiply by zero to get 1)
- The notation $x^{-1}$ means "inverse of $x$" — a pattern you will see again with functions and matrices
