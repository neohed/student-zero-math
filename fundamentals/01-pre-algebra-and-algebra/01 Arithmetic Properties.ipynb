# Fundamental Properties of Arithmetic

Three rules govern how numbers combine under addition and multiplication: **commutativity**, **associativity**, and **distributivity**.

These aren't just abstract laws. They're the reason you can reorder terms, regroup expressions, and factor out common values without changing the result. Compilers use them to optimise machine code. Algebra uses them to solve equations. Understanding them means you can manipulate expressions with confidence — and know exactly when you *can't*.

**Prerequisites:** Basic Python. Basic arithmetic.

---

## Commutativity: Order Doesn't Matter

Some operations don't care about the order of their arguments. You already know this from code:

```python
max(3, 5) == max(5, 3)  # True — max is commutative
```

Swap the arguments, same output. Addition and multiplication work the same way.

But compare that with:

```python
"hello" + " world"  # 'hello world'
" world" + "hello"  # ' worldhello'
```

String concatenation is **not** commutative — order matters. Subtraction and division are the same: swap the operands and you get a different answer.

````python
a, b = 7, 3

print("Addition:")
print(f"  {a} + {b} = {a + b}")
print(f"  {b} + {a} = {b + a}")
print(f"  Commutative: {a + b == b + a}")

print("\nMultiplication:")
print(f"  {a} * {b} = {a * b}")
print(f"  {b} * {a} = {b * a}")
print(f"  Commutative: {a * b == b * a}")

print("\nSubtraction:")
print(f"  {a} - {b} = {a - b}")
print(f"  {b} - {a} = {b - a}")
print(f"  Commutative: {a - b == b - a}")

print("\nDivision:")
print(f"  {a} / {b} = {a / b:.4f}")
print(f"  {b} / {a} = {b / a:.4f}")
print(f"  Commutative: {a / b == b / a}")
````

---

## Associativity: Grouping Doesn't Matter

When you chain the *same* operation, it doesn't matter how you group the pairs.

You know this pattern from boolean logic in Python:

```python
(a and b) and c == a and (b and c)  # True — grouping doesn't change the result
```

Addition and multiplication are associative in the same way. But subtraction and division are **not**:

```python
(10 - 3) - 2 == 5
 10 - (3 - 2) == 9  # different!
```

There's a less obvious gotcha worth knowing: **floating-point arithmetic can also break associativity**, even for addition. This isn't a Python bug — it's a consequence of how floats are represented in hardware, and it has real consequences in numerical computing and ML.

````python
a, b, c = 1, 2, 3

print("Addition:")
print(f"  ({a} + {b}) + {c} = {(a + b) + c}")
print(f"   {a} + ({b} + {c}) = {a + (b + c)}")
print(f"  Associative: {(a + b) + c == a + (b + c)}")

print("\nMultiplication:")
print(f"  ({a} * {b}) * {c} = {(a * b) * c}")
print(f"   {a} * ({b} * {c}) = {a * (b * c)}")
print(f"  Associative: {(a * b) * c == a * (b * c)}")

print("\nSubtraction — grouping matters:")
x, y, z = 10, 3, 2
print(f"  ({x} - {y}) - {z} = {(x - y) - z}")
print(f"   {x} - ({y} - {z}) = {x - (y - z)}")
print(f"  Associative: {(x - y) - z == x - (y - z)}")
````

````python
# The floating-point gotcha
# With floats, even addition is not perfectly associative.

a, b, c = 0.1, 0.2, 0.3

result1 = (a + b) + c
result2 = a + (b + c)

print(f"(0.1 + 0.2) + 0.3 = {result1:.20f}")
print(f" 0.1 + (0.2 + 0.3) = {result2:.20f}")
print(f"Equal? {result1 == result2}")
print(f"Difference: {abs(result1 - result2)}")

# Consequence: summing large arrays gives slightly different results
# depending on order — which is why math.fsum and numpy exist.
import math
import numpy as np

vals = [0.1] * 1_000_000
print(f"\nSum of 1,000,000 copies of 0.1 (expected: 100000.0):")
print(f"  sum():       {sum(vals)}")
print(f"  math.fsum(): {math.fsum(vals)}   <- compensated summation")
print(f"  np.sum():    {np.sum(vals)}")
````

## Distributivity: Spreading an Operation Across a Sum

Distributivity connects multiplication and addition. It says you can either:
- Add first, then multiply, **or**
- Multiply each part separately, then add

You use this pattern constantly. Suppose you're calculating the total tax on a list of prices:

```python
# Option 1: multiply the total
tax = rate * (price1 + price2 + price3)

# Option 2: tax each price separately, then sum
tax = rate * price1 + rate * price2 + rate * price3
```

These are identical. Distributivity is why.

Reading it in reverse is called **factoring**: if you see `a*x + a*y`, you can pull out the shared `a` to get `a*(x + y)`. Same value, simpler form. This is the core move in algebra.

````python
a, b, c = 5, 3, 4

expand     = a * (b + c)       # multiply the sum
distribute = a * b + a * c     # distribute, then sum

print(f"a * (b + c) = {a} * ({b} + {c}) = {a} * {b + c} = {expand}")
print(f"a*b + a*c  = {a}*{b} + {a}*{c} = {a*b} + {a*c} = {distribute}")
print(f"Equal: {expand == distribute}")

# The tax example
rate = 0.2
prices = [49.99, 12.50, 7.00]

tax_on_total = rate * sum(prices)
tax_per_item = sum(rate * p for p in prices)

print(f"\nTax on total:  {tax_on_total:.2f}")
print(f"Tax per item:  {tax_per_item:.2f}")
print(f"Same result:   {abs(tax_on_total - tax_per_item) < 1e-10}")
````

## Visualisation

Both commutativity and distributivity of multiplication have clean geometric proofs using rectangle area.

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

a, b = 4, 3

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# a × b
ax = axes[0]
rect = patches.Rectangle((0, 0), a, b, linewidth=1.5,
                          edgecolor='steelblue', facecolor='lightsteelblue')
ax.add_patch(rect)
ax.text(a / 2, b / 2, f'area = {a * b}', ha='center', va='center', fontsize=13)
ax.set_xlim(-0.3, a + 0.3)
ax.set_ylim(-0.3, b + 0.3)
ax.set_xlabel(f'width = {a}', fontsize=11)
ax.set_ylabel(f'height = {b}', fontsize=11)
ax.set_title(f'{a} x {b} = {a * b}', fontsize=12)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)

# b × a (rotated)
ax = axes[1]
rect = patches.Rectangle((0, 0), b, a, linewidth=1.5,
                          edgecolor='darkorange', facecolor='moccasin')
ax.add_patch(rect)
ax.text(b / 2, a / 2, f'area = {b * a}', ha='center', va='center', fontsize=13)
ax.set_xlim(-0.3, b + 0.3)
ax.set_ylim(-0.3, a + 0.3)
ax.set_xlabel(f'width = {b}', fontsize=11)
ax.set_ylabel(f'height = {a}', fontsize=11)
ax.set_title(f'{b} x {a} = {b * a}', fontsize=12)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)

fig.suptitle('Commutativity: same area, rotated rectangle', fontsize=13)
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

a, b, c = 5, 3, 4

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: a × (b + c) — one big rectangle
ax = axes[0]
rect = patches.Rectangle((0, 0), b + c, a, linewidth=1.5,
                          edgecolor='steelblue', facecolor='lightsteelblue')
ax.add_patch(rect)
ax.text((b + c) / 2, a / 2, f'{a} x {b + c} = {a * (b + c)}',
        ha='center', va='center', fontsize=12)
ax.set_xlim(-0.3, b + c + 0.3)
ax.set_ylim(-0.3, a + 0.3)
ax.set_xlabel(f'b + c = {b} + {c} = {b + c}', fontsize=11)
ax.set_ylabel(f'a = {a}', fontsize=11)
ax.set_title(f'a x (b + c)', fontsize=12)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)

# Right: a×b + a×c — two rectangles side by side
ax = axes[1]
rect1 = patches.Rectangle((0, 0), b, a, linewidth=1.5,
                           edgecolor='steelblue', facecolor='lightsteelblue')
rect2 = patches.Rectangle((b, 0), c, a, linewidth=1.5,
                           edgecolor='darkorange', facecolor='moccasin')
ax.add_patch(rect1)
ax.add_patch(rect2)
ax.text(b / 2, a / 2, f'{a}x{b}={a * b}', ha='center', va='center', fontsize=11)
ax.text(b + c / 2, a / 2, f'{a}x{c}={a * c}', ha='center', va='center', fontsize=11)
ax.set_xlim(-0.3, b + c + 0.3)
ax.set_ylim(-0.3, a + 0.3)
ax.set_xlabel(f'b = {b}    |    c = {c}', fontsize=11)
ax.set_ylabel(f'a = {a}', fontsize=11)
ax.set_title(f'a*b + a*c = {a * b} + {a * c} = {a * b + a * c}', fontsize=12)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)

fig.suptitle('Distributivity: one big rectangle = two smaller rectangles', fontsize=13)
plt.tight_layout()
plt.show()
````

## Formal Notation

Now that the concepts are built, here is the compact mathematical shorthand:

**Commutativity**
$$a + b = b + a \qquad a \times b = b \times a$$

**Associativity**
$$(a + b) + c = a + (b + c) \qquad (a \times b) \times c = a \times (b \times c)$$

**Distributivity**
$$a \times (b + c) = a \times b + a \times c$$

And in reverse (factoring):
$$a \times b + a \times c = a \times (b + c)$$

These hold for all real numbers $a$, $b$, $c$. Subtraction and division satisfy **neither** commutativity nor associativity.

---

## Real-World Connection

These properties appear throughout computing:

- **Compiler optimisations**: compilers reorder and regroup arithmetic based on commutativity and associativity. The `-ffast-math` flag in C/C++ compilers tells the compiler to apply these rules even for floats — which can change results slightly, but enables vectorisation and other speedups.
- **Parallel summation**: to sum an array across multiple CPU cores, you split it into chunks and sum each independently. Associativity is what guarantees the partial sums can be combined in any order and still produce the correct total.
- **Neural network layers**: matrix multiplication is **not** commutative — `A @ B != B @ A`. When writing a forward pass as `output = weights @ input + bias`, the order of operands is not arbitrary. Knowing which operations are commutative (and which aren't) tells you what you can safely rearrange.
- **Symbolic algebra** (SymPy, Mathematica): these tools use commutativity, associativity, and distributivity as rewrite rules to simplify expressions automatically.

---

## Summary

- **Commutativity** — swap the operands, same result: `a + b = b + a`, `a * b = b * a`. Subtraction and division don't have this.
- **Associativity** — change the grouping, same result: `(a + b) + c = a + (b + c)`. Subtraction and division don't have this. Neither do floats, precisely — grouping affects rounding.
- **Distributivity** — multiplication spreads over addition: `a * (b + c) = a*b + a*c`. Reading it right-to-left is factoring.

These rules define which transformations of an expression are *safe* — they change the form without changing the value. That is the foundation of all algebraic manipulation.
