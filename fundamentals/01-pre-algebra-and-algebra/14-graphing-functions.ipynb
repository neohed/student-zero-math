# Graphing Functions

A graph is not decoration — it is a precise mathematical object. Every point $(x, f(x))$ on the curve is an (input, output) pair. The graph traces the behaviour of the function across its entire domain in one image.

This notebook covers how to read and draw function graphs, the visual test for whether a curve represents a function, and the characteristic shapes of common function families.

**Prerequisites:** Notebook 13 (Functions) and Notebook 08 (Geometry and Lines).

---

## 1. What a Graph Shows

To graph $f(x)$: for every $x$ in the domain, compute $f(x)$ and plot the point $(x, f(x))$. Connect the points.

In code, this is literally what matplotlib does when you call `ax.plot(x_values, y_values)`. The programmer's mental model is exactly right: a graph is a scatter plot of every (input, output) pair, drawn densely enough to look like a curve.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')

def f(x):
    return x**2 - 2*x - 3

# Show what a graph is: a collection of (x, f(x)) points
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: sparse points to make the structure visible
ax = axes[0]
x_sparse = np.linspace(-2, 4, 15)
y_sparse = [f(x) for x in x_sparse]
ax.scatter(x_sparse, y_sparse, color='#4c9be8', s=60, zorder=5)
for xi, yi in zip(x_sparse, y_sparse):
    ax.plot([xi, xi], [0, yi], color='#4c9be8', alpha=0.2, linewidth=1)
ax.axhline(y=0, color='#666666', linewidth=0.8)
ax.axvline(x=0, color='#666666', linewidth=0.8)
ax.set_title('f(x) = x\u00b2 \u2212 2x \u2212 3  \u2014  sparse points', fontsize=11)
ax.set_xlabel('x (input)', fontsize=10)
ax.set_ylabel('f(x) (output)', fontsize=10)
ax.grid(alpha=0.15)
ax.text(0.5, 0.04, 'Each point is one (input, output) pair',
        transform=ax.transAxes, ha='center', fontsize=9, color='#aaaaaa')

# Right: dense points => continuous-looking curve
ax = axes[1]
x_dense = np.linspace(-2, 4, 400)
y_dense = [f(x) for x in x_dense]
ax.plot(x_dense, y_dense, color='#4c9be8', linewidth=2.5)
ax.axhline(y=0, color='#666666', linewidth=0.8)
ax.axvline(x=0, color='#666666', linewidth=0.8)
ax.set_title('Same function \u2014 400 points, looks continuous', fontsize=11)
ax.set_xlabel('x (input)', fontsize=10)
ax.set_ylabel('f(x) (output)', fontsize=10)
ax.grid(alpha=0.15)
ax.text(0.5, 0.04, 'Same data, plotted densely',
        transform=ax.transAxes, ha='center', fontsize=9, color='#aaaaaa')

plt.suptitle('A graph is a dense scatter plot of (x, f(x)) pairs', fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

## 2. The Vertical Line Test

A curve on a graph represents a function if and only if every vertical line crosses it **at most once**.

The reasoning is direct: a vertical line at $x = a$ hits the curve at every output value for that input. If it hits twice, that input produces two outputs — which violates the definition of a function.

This is the graphical version of the same rule from Notebook 13: one input, one output.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Left: y = x^2 (IS a function) ---
ax = axes[0]
x = np.linspace(-3, 3, 300)
ax.plot(x, x**2, color='#56b4a0', linewidth=2.5)
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-0.5, 9.5)
ax.axhline(y=0, color='#555555', linewidth=0.8)
ax.axvline(x=0, color='#555555', linewidth=0.8)
ax.grid(alpha=0.15)
ax.set_title('y = x\u00b2   \u2714 function', fontsize=12, color='#56b4a0')

for xi in [-2.0, 0.5, 2.0]:
    yi = xi**2
    ax.axvline(x=xi, color='#ffd700', linewidth=1.5, linestyle='--', alpha=0.8)
    ax.scatter([xi], [yi], color='#ffd700', s=70, zorder=6)

ax.text(0.5, 0.06, 'Each vertical line crosses once \u2192 one output per input',
        transform=ax.transAxes, ha='center', fontsize=9, color='#aaaaaa')

# --- Right: circle x^2 + y^2 = 1 (NOT a function) ---
ax = axes[1]
t = np.linspace(0, 2*np.pi, 400)
ax.plot(np.cos(t), np.sin(t), color='#e8914c', linewidth=2.5)
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)
ax.set_aspect('equal')
ax.axhline(y=0, color='#555555', linewidth=0.8)
ax.axvline(x=0, color='#555555', linewidth=0.8)
ax.grid(alpha=0.15)
ax.set_title('x\u00b2 + y\u00b2 = 1   \u2718 not a function', fontsize=12, color='#e8914c')

for xi in [-0.6, 0.0, 0.8]:
    if abs(xi) <= 1:
        y_top =  np.sqrt(1 - xi**2)
        y_bot = -np.sqrt(1 - xi**2)
        ax.axvline(x=xi, color='#ffd700', linewidth=1.5, linestyle='--', alpha=0.8)
        ax.scatter([xi, xi], [y_top, y_bot], color='#ffd700', s=70, zorder=6)

ax.text(0.5, 0.06, 'Vertical lines cross twice \u2192 two outputs for one input',
        transform=ax.transAxes, ha='center', fontsize=9, color='#aaaaaa')

plt.suptitle('The Vertical Line Test', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

## 3. Linear Functions

A linear function produces a straight line:

$$f(x) = mx + b$$

- $m$ is the **slope** — the rate of change. For every 1 unit increase in $x$, the output increases by $m$.
- $b$ is the **y-intercept** — the output when $x = 0$.

This is the same slope-intercept form from Notebook 08, now viewed as a function. The connection is direct: a straight line on a graph is the visual trace of a linear function.

Slope encodes the relationship: positive slope rises left to right, negative slope falls, zero slope is flat (constant function).

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

x = np.linspace(-4, 4, 300)

# --- Left: vary slope, fix intercept ---
ax = axes[0]
slopes = [(-2, '#e8914c'), (-0.5, '#e87070'), (0, '#aaaaaa'), (1, '#56b4a0'), (3, '#4c9be8')]
b = 1
for m, color in slopes:
    label = f'm={m}, b={b}'
    ax.plot(x, m*x + b, color=color, linewidth=2, label=label)
ax.axhline(y=0, color='#555555', linewidth=0.8)
ax.axvline(x=0, color='#555555', linewidth=0.8)
ax.set_xlim(-4, 4)
ax.set_ylim(-8, 12)
ax.grid(alpha=0.15)
ax.legend(fontsize=9, loc='upper left')
ax.set_title('Varying slope m  (b = 1)', fontsize=11)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# --- Right: vary intercept, fix slope ---
ax = axes[1]
m = 1
intercepts = [(-3, '#e8914c'), (-1, '#e87070'), (0, '#aaaaaa'), (2, '#56b4a0'), (4, '#4c9be8')]
for b, color in intercepts:
    ax.plot(x, m*x + b, color=color, linewidth=2, label=f'm={m}, b={b}')
ax.axhline(y=0, color='#555555', linewidth=0.8)
ax.axvline(x=0, color='#555555', linewidth=0.8)
ax.set_xlim(-4, 4)
ax.set_ylim(-8, 9)
ax.grid(alpha=0.15)
ax.legend(fontsize=9, loc='upper left')
ax.set_title('Varying intercept b  (m = 1)', fontsize=11)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

plt.suptitle('f(x) = mx + b  \u2014  slope controls tilt, intercept controls vertical position', fontsize=12, y=1.01)
plt.tight_layout()
plt.show()
```

## 4. Non-Linear Functions

Not all functions are straight lines. The shape of the graph reveals the function's behaviour — growth rate, turning points, asymptotes, and domain restrictions.

Four functions worth knowing by their shapes:

| Function | Name | Key behaviour |
|----------|------|---------------|
| $f(x) = x^2$ | Quadratic / Parabola | U-shape, symmetric, minimum at vertex |
| $f(x) = x^3$ | Cubic | S-shape, through origin, no bounds |
| $f(x) = \frac{1}{x}$ | Reciprocal / Hyperbola | Asymptotes at $x=0$ and $y=0$ |
| $f(x) = \sqrt{x}$ | Square root | Defined only for $x \geq 0$, decelerating growth |

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')

fig, axes = plt.subplots(2, 2, figsize=(12, 9))

def add_axes(ax):
    ax.axhline(y=0, color='#555555', linewidth=0.8)
    ax.axvline(x=0, color='#555555', linewidth=0.8)
    ax.grid(alpha=0.15)

# Quadratic
ax = axes[0][0]
x = np.linspace(-3, 3, 300)
ax.plot(x, x**2, color='#4c9be8', linewidth=2.5)
add_axes(ax)
ax.set_title('f(x) = x\u00b2  (quadratic)', fontsize=11)
ax.set_ylim(-0.5, 9)
ax.text(0.05, 0.88, 'U-shape, symmetric\nminimum at (0, 0)',
        transform=ax.transAxes, fontsize=9, color='#aaaaaa')

# Cubic
ax = axes[0][1]
ax.plot(x, x**3, color='#56b4a0', linewidth=2.5)
add_axes(ax)
ax.set_title('f(x) = x\u00b3  (cubic)', fontsize=11)
ax.set_ylim(-10, 10)
ax.text(0.05, 0.88, 'S-shape, rotational symmetry\nthrough origin',
        transform=ax.transAxes, fontsize=9, color='#aaaaaa')

# Reciprocal: plot two branches separately to avoid the vertical line through zero
ax = axes[1][0]
x_neg = np.linspace(-4, -0.1, 200)
x_pos = np.linspace(0.1, 4, 200)
ax.plot(x_neg, 1/x_neg, color='#e8914c', linewidth=2.5)
ax.plot(x_pos, 1/x_pos, color='#e8914c', linewidth=2.5)
add_axes(ax)
ax.set_xlim(-4, 4)
ax.set_ylim(-6, 6)
ax.set_title('f(x) = 1/x  (reciprocal)', fontsize=11)
ax.text(0.05, 0.88, 'Asymptotes at x=0, y=0\ndomain: x \u2260 0',
        transform=ax.transAxes, fontsize=9, color='#aaaaaa')

# Square root
ax = axes[1][1]
x_sq = np.linspace(0, 9, 300)
ax.plot(x_sq, np.sqrt(x_sq), color='#a78bde', linewidth=2.5)
add_axes(ax)
ax.set_xlim(-0.5, 9)
ax.set_ylim(-0.5, 3.5)
ax.set_title('f(x) = \u221ax  (square root)', fontsize=11)
ax.text(0.05, 0.88, 'Domain: x \u2265 0\ngrowth decelerates',
        transform=ax.transAxes, fontsize=9, color='#aaaaaa')

plt.suptitle('Common non-linear function shapes', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

## 5. Piecewise Functions

A piecewise function applies different rules in different regions of its domain. In mathematics this is written with a brace:

$$f(x) = \begin{cases} -x & \text{if } x < 0 \\ x^2 & \text{if } 0 \leq x < 2 \\ 4 & \text{if } x \geq 2 \end{cases}$$

In Python, this is an if/else block. The mathematician's notation and the programmer's intuition are identical — this is just a function with conditional logic.

Piecewise functions appear throughout engineering and physics wherever behaviour changes depending on conditions: a tax bracket formula, a piecewise linear approximation, a control system that behaves differently in different regimes.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    if x < 0:
        return -x        # reflect: same as abs() for negatives
    elif x < 2:
        return x**2      # quadratic in the middle region
    else:
        return 4         # flat from x=2 onward

print('f(x) defined piecewise:')
print('  -x      for x < 0')
print('  x^2     for 0 <= x < 2')
print('  4       for x >= 2')
print()
print(f'  {"x":>5}  {"f(x)":>6}  region')
print(f'  {"-"*5}  {"-"*6}  {"-"*15}')
for x in [-3, -1, 0, 0.5, 1, 1.5, 2, 3, 5]:
    fx = f(x)
    region = 'x < 0' if x < 0 else ('0 <= x < 2' if x < 2 else 'x >= 2')
    print(f'  {x:>5}  {fx:>6.2f}  {region}')

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(9, 5))

x_vals = np.linspace(-3.5, 5, 600)
y_vals = [f(x) for x in x_vals]

ax.plot(x_vals, y_vals, color='#4c9be8', linewidth=2.5)
ax.axhline(y=0, color='#555555', linewidth=0.8)
ax.axvline(x=0, color='#555555', linewidth=0.8)

# Mark the transition points
for bx, by in [(0, f(0)), (2, f(2))]:
    ax.scatter([bx], [by], color='#ffd700', s=80, zorder=6)
    ax.axvline(x=bx, color='#ffd700', linewidth=1, linestyle=':', alpha=0.5)

# Annotate each region
ax.text(-2, 1.5, '-x', fontsize=12, color='#aaaaaa', ha='center')
ax.text(1, 1.5, 'x\u00b2', fontsize=12, color='#aaaaaa', ha='center')
ax.text(3.5, 4.4, '4', fontsize=12, color='#aaaaaa', ha='center')

ax.set_xlim(-3.5, 5)
ax.set_ylim(-0.5, 6)
ax.set_xlabel('x', fontsize=11)
ax.set_ylabel('f(x)', fontsize=11)
ax.set_title('Piecewise function  \u2014  different rules in different regions', fontsize=12)
ax.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

## 6. Transformations

Modifying the function formula moves or stretches the graph in predictable ways. The rules are worth knowing because they let you reason about the shape of a function without plotting it.

Starting from a base function $f(x)$:

| Transformation | Effect on graph |
|---------------|----------------|
| $f(x) + k$ | Shift up by $k$ (down if $k < 0$) |
| $f(x - h)$ | Shift right by $h$ (left if $h < 0$) |
| $a \cdot f(x)$ | Stretch vertically by factor $a$ (compress if $0 < a < 1$) |
| $f(ax)$ | Compress horizontally by factor $a$ |
| $-f(x)$ | Reflect over the x-axis |

The vertical shift is intuitive: adding a constant to the output raises every point. The horizontal shift has a counterintuitive sign: $f(x - 3)$ shifts the graph **right** by 3, because you need $x = 3$ to get the same input value as $x = 0$ in the original.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')

x = np.linspace(-5, 5, 400)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# --- Vertical shifts ---
ax = axes[0]
for k, color, label in [
    (0,   '#666666',  'f(x) = x\u00b2  (base)'),
    (3,   '#4c9be8',  'f(x) + 3  (up 3)'),
    (-2,  '#e8914c',  'f(x) \u2212 2  (down 2)'),
]:
    ax.plot(x, x**2 + k, color=color, linewidth=2, label=label)
ax.set_ylim(-3, 16)
ax.set_xlim(-4, 4)
ax.axhline(y=0, color='#444444', linewidth=0.8)
ax.axvline(x=0, color='#444444', linewidth=0.8)
ax.grid(alpha=0.15)
ax.legend(fontsize=8)
ax.set_title('Vertical shifts  f(x) + k', fontsize=11)

# --- Horizontal shifts ---
ax = axes[1]
for h, color, label in [
    (0,   '#666666',  'f(x) = x\u00b2  (base)'),
    (3,   '#56b4a0',  'f(x\u22123)  (right 3)'),
    (-2,  '#e87070',  'f(x+2)  (left 2)'),
]:
    ax.plot(x, (x - h)**2, color=color, linewidth=2, label=label)
ax.set_ylim(-0.5, 12)
ax.set_xlim(-5, 5)
ax.axhline(y=0, color='#444444', linewidth=0.8)
ax.axvline(x=0, color='#444444', linewidth=0.8)
ax.grid(alpha=0.15)
ax.legend(fontsize=8)
ax.set_title('Horizontal shifts  f(x \u2212 h)', fontsize=11)

# --- Vertical stretch / reflect ---
ax = axes[2]
for a, color, label in [
    (1,   '#666666',  'f(x) = x\u00b2  (base)'),
    (2.5, '#4c9be8',  '2.5 \u00b7 f(x)  (stretch)'),
    (0.4, '#56b4a0',  '0.4 \u00b7 f(x)  (compress)'),
    (-1,  '#e8914c',  '\u2212f(x)  (reflect)'),
]:
    ax.plot(x, a * x**2, color=color, linewidth=2, label=label)
ax.set_ylim(-10, 22)
ax.set_xlim(-3, 3)
ax.axhline(y=0, color='#444444', linewidth=0.8)
ax.axvline(x=0, color='#444444', linewidth=0.8)
ax.grid(alpha=0.15)
ax.legend(fontsize=8)
ax.set_title('Vertical stretch  a \u00b7 f(x)', fontsize=11)

plt.suptitle('Transformations of f(x) = x\u00b2', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

## Formal Notation

**Vertical line test:** A curve in the plane represents a function if and only if every vertical line $x = a$ intersects the curve at most once.

**Linear function:** $f(x) = mx + b$, where $m$ is the slope and $b$ is the y-intercept. The graph is a line with gradient $m$ crossing the y-axis at $(0, b)$.

**Transformations** of $y = f(x)$:
- $y = f(x) + k$: vertical translation by $k$
- $y = f(x - h)$: horizontal translation by $h$ (positive $h$ moves right)
- $y = a \cdot f(x)$: vertical scaling by factor $a$; $a < 0$ reflects over x-axis
- $y = f(ax)$: horizontal scaling; $a > 1$ compresses, $0 < a < 1$ stretches

**Piecewise function:** a function defined by multiple sub-functions, each valid on a sub-interval of the domain. The notation specifies both the rule and the interval for each piece.

## Real-World Connection

- **Rendering and graphics**: every pixel on screen is a function evaluation. A shader is a function $f(x, y) \to$ colour. The entire image is the graph of that function over a 2D domain.
- **Machine learning activation functions**: ReLU is a piecewise function: $f(x) = 0$ for $x < 0$, $f(x) = x$ for $x \geq 0$. Its shape — flat then linear — is why it is graphed in every neural network introduction. Understanding the graph is understanding the function's effect on gradients.
- **Control systems and signals**: a transfer function maps input signals to output signals. Engineers read the shape of the frequency-response graph to understand stability, gain, and phase shift without examining every input.
- **Game physics**: collision detection, gravity, and force functions are all non-linear. Knowing the shape of $1/r^2$ (inverse square law) tells you immediately that gravity weakens fast with distance — the graph makes this visceral.
- **Transformations in data preprocessing**: normalisation is a vertical shift and scale: $(x - \mu) / \sigma$. Log-transforming skewed data applies $f(x) = \log(x)$. These are function transformations applied to every data point.

---

## Summary

- A **graph** is a dense plot of $(x, f(x))$ pairs. It is a visual representation of the function's complete behaviour over its domain.
- The **vertical line test**: a curve is a function if every vertical line crosses it at most once. One crossing = one output per input.
- **Linear functions** $f(x) = mx + b$ produce straight lines. Slope controls tilt; intercept controls vertical position.
- **Non-linear functions** have characteristic shapes: parabola ($x^2$), S-curve ($x^3$), hyperbola ($1/x$), decelerating curve ($\sqrt{x}$). Knowing these shapes lets you read the function's behaviour at a glance.
- **Piecewise functions** use different rules in different regions. In code, this is an if/else block. The mathematician's brace notation and the programmer's conditional are the same thing.
- **Transformations** move and stretch the graph predictably: $+k$ shifts up, $-h$ inside the argument shifts right, $a \cdot$ stretches vertically, $-f(x)$ reflects. These rules let you sketch a transformed function without recomputing every point.
