# Cross Multiplication: Solving Proportions

Cross multiplication is a technique for solving **proportions** — equations where one fraction equals another and one value is unknown.

It looks like a trick, but it isn't. It's just clearing the denominators in one move. This notebook builds the intuition, derives the rule, and shows when to reach for it.

**Prerequisites:** Notebooks 01–05.

---

## Intuition: Proportions as Scaling

A proportion is a statement that two ratios are equal. You deal with this constantly in code:

```python
# Recipe scaling: 3 cups makes 24 cookies.
# How many cups do you need for 60 cookies?
#
#   3 / 24 = x / 60
#
# The ratio of cups to cookies stays the same — the recipe just scales.

# Image resizing: original is 800 x 600.
# Scale to width 1200. What's the new height?
#
#   800 / 600 = 1200 / h
#
# The aspect ratio is preserved.

# Unit conversion: 1 mile = 1.609 km.
# How many km is 7.5 miles?
#
#   1 / 1.609 = 7.5 / km
```

In each case the shape is the same: two equal fractions, one unknown. Cross multiplication solves for that unknown.

---

## Where the Rule Comes From

Start with a proportion:

$$\frac{a}{b} = \frac{c}{d}$$

Multiply both sides by $b$ to clear the left denominator:

$$a = \frac{bc}{d}$$

Multiply both sides by $d$ to clear the right denominator:

$$ad = bc$$

That's it. Cross multiplication is just **multiplying both sides by both denominators** in one step. The name comes from the visual pattern: $a$ and $d$ are diagonally opposite, as are $b$ and $c$.

````python
from fractions import Fraction

def solve_proportion(a, b, c, d, unknown):
    """
    Solve a proportion a/b = c/d for one unknown.
    Pass None for the unknown value.
    'unknown' must be 'a', 'b', 'c', or 'd'.

    From ad = bc:
      a = bc/d
      b = ad/c
      c = ad/b
      d = bc/a
    """
    if unknown == 'a': result = Fraction(b * c, d)
    elif unknown == 'b': result = Fraction(a * d, c)
    elif unknown == 'c': result = Fraction(a * d, b)
    elif unknown == 'd': result = Fraction(b * c, a)

    vals = {'a': a, 'b': b, 'c': c, 'd': d}
    vals[unknown] = result
    a2, b2, c2, d2 = vals['a'], vals['b'], vals['c'], vals['d']

    print(f"  Proportion:  {a}/{b} = {c}/{d}  (? = {unknown})")
    print(f"  Cross mult:  {a} * {d} = {b} * {c}")
    print(f"  Solved:      {unknown} = {result}")
    print(f"  Verify:      {a2}/{b2} = {Fraction(a2, b2)}   ==   {c2}/{d2} = {Fraction(c2, d2)}  =>  {Fraction(a2, b2) == Fraction(c2, d2)}")
    print()
    return result

# Example 1: x/3 = 4/5
solve_proportion('x', 3, 4, 5, 'a')

# Example 2: 2/x = 5/7
solve_proportion(2, 'x', 5, 7, 'b')
````

The function above is more for illustration. In practice, with concrete numbers, Python's `Fraction` does the algebra cleanly:

````python
from fractions import Fraction

def solve_proportion(a, b, c, d, unknown):
    """
    Solve a/b = c/d for the named unknown (pass 0 as placeholder).
    Uses cross multiplication: ad = bc.
    """
    if   unknown == 'a': return Fraction(b * c, d)
    elif unknown == 'b': return Fraction(a * d, c)
    elif unknown == 'c': return Fraction(a * d, b)
    elif unknown == 'd': return Fraction(b * c, a)

# Recipe: 3 cups / 24 cookies = x cups / 60 cookies
cups = solve_proportion(3, 24, 0, 60, 'c')
print(f"Recipe scaling:   3/24 = x/60  =>  x = {cups} cups")

# Image resize: 800/600 = 1200/h
height = solve_proportion(800, 600, 1200, 0, 'd')
print(f"Image resize:     800/600 = 1200/h  =>  h = {height}px")

# Unit conversion: 1/1.609 = 7.5/km  (using integers: 1000/1609 = 7500/km)
km = solve_proportion(1000, 1609, 7500, 0, 'd')
print(f"Miles to km:      7.5 miles = {float(km):.3f} km")
````

## Visualisation

### The Cross Pattern

The name comes from the arrows drawn on the proportion. When $\frac{a}{b} = \frac{c}{d}$, you multiply across the diagonals:

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyArrowPatch

def plot_cross_multiplication(a, b, c, d, unknown_label='x'):
    fig, ax = plt.subplots(figsize=(7, 3.5))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 4)
    ax.axis('off')

    # Labels
    font = dict(fontsize=22, ha='center', va='center')
    ax.text(2, 3, str(a), **font, color='steelblue')
    ax.text(2, 1, str(b), **font, color='steelblue')
    ax.text(8, 3, str(c), **font, color='darkorange')
    ax.text(8, 1, str(d), **font, color='darkorange')

    # Fraction bars
    ax.plot([1.2, 2.8], [2, 2], 'black', linewidth=1.5)
    ax.plot([7.2, 8.8], [2, 2], 'black', linewidth=1.5)

    # Equals sign
    ax.text(5, 2, '=', fontsize=24, ha='center', va='center', color='black')

    # Cross arrows
    arrow_kw = dict(arrowstyle='->', lw=1.5, connectionstyle='arc3,rad=0.0')
    ax.annotate('', xy=(7.8, 0.8), xytext=(2.2, 3.2),
                arrowprops=dict(**arrow_kw, color='green'))
    ax.annotate('', xy=(7.8, 3.2), xytext=(2.2, 0.8),
                arrowprops=dict(**arrow_kw, color='crimson'))

    # Equation below
    ax.text(5, 0.2,
            f'{a} x {d}  =  {b} x {c}',
            fontsize=14, ha='center', va='center', color='gray')

    ax.set_title(f'Cross multiplication: a/b = c/d  =>  a×d = b×c', fontsize=11)
    plt.tight_layout()
    plt.show()

plot_cross_multiplication('x', 3, 4, 5)
````

````python
import matplotlib.pyplot as plt

def plot_proportion(a, b, c, d, labels=('ratio 1', 'ratio 2')):
    """Show two equal ratios as bars scaled to the same proportion."""
    r1 = a / b
    r2 = c / d

    fig, axes = plt.subplots(1, 2, figsize=(10, 3), sharey=True)

    for ax, total, part, label, colour in zip(
        axes, [b, d], [a, c], labels, ['steelblue', 'darkorange']
    ):
        ax.barh([1], [total], color='lightgray', edgecolor='black', height=0.4)
        ax.barh([1], [part],  color=colour,      edgecolor='black', height=0.4, alpha=0.8)
        ax.set_xlim(0, max(b, d) * 1.1)
        ax.set_yticks([])
        ax.set_title(f'{label}\n{part}/{total} = {part/total:.3f}', fontsize=10)
        ax.set_xlabel('value')
        ax.grid(axis='x', alpha=0.3)

    fig.suptitle(f'Proportion: {a}/{b} = {c}/{d}  (both = {r1:.4f})', fontsize=11)
    plt.tight_layout()
    plt.show()

# Recipe: 3/24 = 7.5/60 (cups to cookies)
plot_proportion(3, 24, 7.5, 60, ('3 cups / 24 cookies', '7.5 cups / 60 cookies'))

# Image resize: 800/600 = 1200/900
plot_proportion(800, 600, 1200, 900, ('800 x 600', '1200 x 900'))
````

## A Common Mistake: When NOT to Cross Multiply

Cross multiplication only works when two fractions are **equal to each other** — a proportion.

It does **not** apply when fractions are being added or subtracted:

```
WRONG:  1/3 + 1/4 = ?  ... cross multiplying gives nothing useful here
RIGHT:  use the LCM method (notebook 05)

CORRECT use:  1/3 = x/12  ... solve for x using cross multiplication
```

The test: does your equation have the form `fraction = fraction`? If yes, cross multiply. If it involves `+` or `-` between fractions, use LCM.

````python
from fractions import Fraction

# Correct: testing whether two fractions form a true proportion
def is_proportion(a, b, c, d):
    """Return True if a/b == c/d."""
    return a * d == b * c  # cross multiply: ad == bc

tests = [
    (1, 2, 3, 6),    # 1/2 == 3/6 ?
    (2, 3, 4, 5),    # 2/3 == 4/5 ?
    (3, 4, 9, 12),   # 3/4 == 9/12 ?
    (5, 8, 10, 15),  # 5/8 == 10/15 ?
]

for a, b, c, d in tests:
    result = is_proportion(a, b, c, d)
    print(f"  {a}/{b} == {c}/{d} ?  =>  {a}*{d} == {b}*{c}  =>  {a*d} == {b*c}  =>  {result}")
````

## Formal Notation

Given a proportion where $b, d \neq 0$:

$$\frac{a}{b} = \frac{c}{d} \implies ad = bc$$

To solve for any one unknown, isolate it:

$$a = \frac{bc}{d} \qquad b = \frac{ad}{c} \qquad c = \frac{ad}{b} \qquad d = \frac{bc}{a}$$

The derivation — multiply both sides of $\frac{a}{b} = \frac{c}{d}$ by $bd$ — shows this is just algebraic manipulation, not a separate rule.

---

## Real-World Connection

- **Image and UI scaling**: every proportional resize operation is a proportion solved by cross multiplication. `original_width / original_height = new_width / new_height` — solve for the unknown dimension.
- **Unit conversion**: the entire dimensional analysis technique used in physics and engineering is chains of proportions. Converting 60 miles/hour to metres/second involves multiplying by conversion ratios — each step is a proportion.
- **Similarity in geometry**: two triangles are similar if their sides are proportional. Testing similarity and finding unknown side lengths both use cross multiplication.
- **A/B testing**: if 120 out of 400 users clicked in variant A, and the test is scaled to 2,500 users, how many clicks would you expect? `120/400 = x/2500` — cross multiply to find `x`.
- **Odds and ratios in probability**: converting between probability and odds, or scaling frequencies from a sample to a population, are all proportion problems.

---

## Summary

- A proportion is two equal fractions: `a/b = c/d`
- Cross multiplication clears both denominators in one step: `ad = bc`
- It's not a trick — it's multiplying both sides by `bd`
- Solve for any unknown by isolating it: if solving for `c`, then `c = ad/b`
- Use it when a fraction equals another fraction with one unknown — not for adding or subtracting fractions
- `is_proportion(a, b, c, d)` reduces to checking whether `a*d == b*c`
