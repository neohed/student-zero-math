# Functions

A mathematical function takes an input, applies a rule, and produces an output.

That description also fits every function you have ever written in code. This is not a loose analogy — it is the same concept with different notation.

**Prerequisites:** Notebooks 07 (Variables and Equations) and 08 (Geometry and Lines).

---

## 1. The Programmer Bridge

Here is a mathematical function:

$$f(x) = 2x + 3$$

Here is a Python function:

```python
def f(x):
    return 2 * x + 3
```

They are the same thing. Same input variable. Same rule. Same output. The math notation $f(x)$ means exactly what `f(x)` means in Python: call function `f` with argument `x`.

This is the clearest mapping from mathematics to code you will encounter. Every concept in this notebook is something you already understand from programming — the only new thing is notation.

````python
def f(x):
    return 2 * x + 3

print('f(x) = 2x + 3')
print()
for x in range(-3, 4):
    print(f'  f({x:2}) = {f(x)}')

print()
print('One input. One output. The same rule every time.')
print('That is the definition of a function in both mathematics and programming.')
````

## 2. Relations and What Makes a Function

A **relation** is any set of (input, output) pairs — a mapping from some inputs to some outputs. A relation becomes a **function** when every input maps to exactly one output.

The rule: **one input, one output**.

- Multiple inputs can map to the same output — that is allowed.
- One input mapping to multiple outputs — that is not allowed.

This matches the programming intuition exactly. A pure function called with the same argument always returns the same value. If `f(3)` returns 7 today and 11 tomorrow with no change in state, that is a broken function. Mathematics has the same constraint, baked in by definition.

````python
def is_function(pairs):
    """A relation is a function if each input appears at most once."""
    inputs = [p[0] for p in pairs]
    return len(inputs) == len(set(inputs))

relations = [
    ('each input appears once',             [(1, 2), (2, 4), (3, 6)]),
    ('different inputs, same output',        [(1, 2), (2, 2), (3, 2)]),
    ('one input maps to two outputs',        [(1, 2), (1, 3), (2, 4)]),
    ('all inputs map to same output',        [(1, 5), (2, 5), (3, 5)]),
    ('only one pair',                        [(7, 42)]),
]

print(f'  {"Result":<14}  {"Pairs"}')
print(f'  {"-"*14}  {"-"*40}')
for description, pairs in relations:
    result = 'function' if is_function(pairs) else 'NOT function'
    print(f'  {result:<14}  {pairs}')
    print(f'  {"":14}  {description}')
    print()
````

## 3. Domain and Range

Every function has a set of valid inputs and a corresponding set of outputs:

- **Domain** — the set of all valid inputs
- **Range** — the set of all outputs the function can produce

In programming terms: the domain is the set of inputs that do not cause your function to throw. The range is the set of values it can return.

Some functions have restricted domains:

| Function | Restriction | Reason |
|----------|-------------|--------|
| $f(x) = \frac{1}{x}$ | $x \neq 0$ | Division by zero |
| $f(x) = \sqrt{x}$ | $x \geq 0$ | Square root of negative is not real |
| $f(x) = \log(x)$ | $x > 0$ | Logarithm undefined for zero and negatives |
| $f(x) = 2x + 3$ | all real numbers | No restrictions |

````python
import math

def f_reciprocal(x):
    if x == 0:
        raise ValueError('Domain error: x cannot be 0 (division by zero)')
    return 1 / x

def f_sqrt(x):
    if x < 0:
        raise ValueError(f'Domain error: sqrt requires x >= 0 (got {x})')
    return math.sqrt(x)

def f_log(x):
    if x <= 0:
        raise ValueError(f'Domain error: log requires x > 0 (got {x})')
    return math.log(x)

print('Testing domain boundaries:')
print()

for func, name, test_values in [
    (f_reciprocal, '1/x',    [-2, -1, 0, 1, 2]),
    (f_sqrt,       'sqrt(x)', [-1, 0, 1, 4, 9]),
    (f_log,        'log(x)',  [-1, 0, 1, 2, math.e]),
]:
    print(f'  f(x) = {name}')
    for x in test_values:
        try:
            result = func(x)
            print(f'    f({x:.3g}) = {result:.4g}')
        except ValueError as e:
            print(f'    f({x:.3g}) => {e}')
    print()
````

## 4. Multiple Representations

A function is an abstract rule. That rule can be expressed in four different ways — and they all describe the same thing:

1. **Equation** — a compact algebraic rule
2. **Table** — explicit input/output pairs
3. **Graph** — a visual trace of every (input, output) pair
4. **Code** — an executable implementation

Being comfortable switching between these representations is one of the most useful skills in applied mathematics. A graph shows shape and behaviour at a glance. A table shows exact values. An equation shows the rule in its most general form. Code makes it runnable.

````python
import matplotlib.pyplot as plt
import numpy as np

# The function: f(x) = x^2 + 1

# --- Representation 1: Code ---
def f(x):
    return x**2 + 1

# --- Representation 2: Table ---
print('Equation: f(x) = x^2 + 1')
print()
print('Table:')
print(f'  {"x":>4}  {"f(x)":>6}')
print(f'  {"-"*4}  {"-"*6}')
for x in range(-4, 5):
    print(f'  {x:>4}  {f(x):>6}')

# --- Representation 3: Graph ---
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(7, 5))

x_vals = np.linspace(-4, 4, 300)
y_vals = [f(x) for x in x_vals]

ax.plot(x_vals, y_vals, color='#4c9be8', linewidth=2.5, label='f(x) = x² + 1')
ax.axhline(y=0, color='#666666', linewidth=0.8)
ax.axvline(x=0, color='#666666', linewidth=0.8)

# Mark a few table points
for x in [-3, -2, -1, 0, 1, 2, 3]:
    ax.scatter([x], [f(x)], color='#ffd700', s=40, zorder=5)

ax.set_xlabel('x (input)', fontsize=11)
ax.set_ylabel('f(x) (output)', fontsize=11)
ax.set_title('f(x) = x\u00b2 + 1  \u2014  four representations of the same function', fontsize=12)
ax.legend(fontsize=10)
ax.grid(alpha=0.2)
plt.tight_layout()
plt.show()
````

## 5. Function Notation

$f(x)$ is the standard way to write a function call in mathematics. $f$ is the name, $x$ is the input.

This notation extends naturally:

- $f(3)$ — evaluate the function at $x = 3$
- $f(a + b)$ — evaluate at input $a + b$ (substitute the whole expression)
- $f(2x)$ — evaluate at input $2x$
- $g(x)$, $h(x)$ — different function names, same idea

The common mistake: treating $f(x+1)$ as $f(x) + 1$. It is not. Substitution replaces every occurrence of $x$ with the full expression.

For $f(x) = x^2 + 1$:

$$f(x + 1) = (x+1)^2 + 1 = x^2 + 2x + 1 + 1 = x^2 + 2x + 2$$

Not $f(x) + 1 = x^2 + 2$.

````python
def f(x):
    return x**2 + 1

# Evaluating at specific values
print('f(x) = x^2 + 1')
print()
print('Evaluating at specific inputs:')
for val in [0, 3, -2, 0.5]:
    print(f'  f({val}) = {val}^2 + 1 = {f(val)}')

print()
print('Substituting expressions:')
print('  f(x+1) substitutes the whole expression (x+1) for x:')
for x in range(-2, 3):
    direct = f(x + 1)          # f evaluated at (x+1)
    wrong  = f(x) + 1          # common mistake: f(x) + 1
    print(f'  x={x:2}: f({x}+1) = f({x+1}) = {direct}   vs f({x})+1 = {wrong}   same? {direct == wrong}')

print()
print('They only agree when x = 0, by coincidence.')
````

## 6. Composition

**Composition** is applying one function to the output of another:

$$f(g(x))$$

Apply $g$ first, feed the result into $f$. In Python this is just nested function calls — you have been doing this since day one:

```python
result = f(g(x))
```

The mathematical notation $f \circ g$ (read "f composed with g") means the same thing. $(f \circ g)(x) = f(g(x))$.

**Order matters.** $f(g(x))$ and $g(f(x))$ are generally not the same — applying $g$ first then $f$ is a different operation than $f$ first then $g$. This is like the difference between `json.dumps(sorted(data))` and `sorted(json.dumps(data))` — entirely different results.

````python
# Composition is just nested function calls
def f(x):
    return x + 1     # add 1

def g(x):
    return 2 * x     # double

def compose(outer, inner):
    """Returns a new function that applies inner first, then outer."""
    return lambda x: outer(inner(x))

fog = compose(f, g)   # f(g(x)): double first, then add 1 => 2x + 1
gof = compose(g, f)   # g(f(x)): add 1 first, then double => 2(x+1) = 2x + 2

print('f(x) = x + 1')
print('g(x) = 2x')
print()
print('f(g(x)):  double first, then add 1  =>  2x + 1')
print('g(f(x)):  add 1 first, then double  =>  2x + 2')
print()
print(f'  {"x":>3}  {"f(g(x))":>8}  {"g(f(x))":>8}  {"same?":>6}')
print(f'  {"-"*3}  {"-"*8}  {"-"*8}  {"-"*6}')
for x in range(-3, 4):
    fgx = fog(x)
    gfx = gof(x)
    print(f'  {x:>3}  {fgx:>8}  {gfx:>8}  {str(fgx == gfx):>6}')

print()
print('They differ by 2 for every x. Order is not interchangeable.')
````

## 7. Inverse Functions

The inverse of a function undoes it. If $f$ maps $x$ to $y$, then $f^{-1}$ maps $y$ back to $x$:

$$f^{-1}(f(x)) = x$$

Programmer bridge: encode/decode, serialize/deserialize, encrypt/decrypt — these are all inverse function pairs.

**Finding an inverse algebraically** uses the equation manipulation from Notebook 07:

1. Write the function as $y = ...$
2. Swap $x$ and $y$
3. Solve for $y$

For $f(x) = 2x + 3$:
$$y = 2x + 3 \quad\Rightarrow\quad x = 2y + 3 \quad\Rightarrow\quad y = \frac{x - 3}{2}$$

So $f^{-1}(x) = \frac{x-3}{2}$.

**Not every function has a unique inverse.** $f(x) = x^2$ maps both $4$ and $-4$ to $16$ — so $f^{-1}(16)$ is ambiguous. To invert it you must restrict the domain.

````python
import math

# f(x) = 2x + 3,  f_inv(x) = (x - 3) / 2
def f(x):
    return 2 * x + 3

def f_inv(x):
    return (x - 3) / 2

print('f(x)     = 2x + 3')
print('f_inv(x) = (x - 3) / 2')
print()
print('Verification: f_inv(f(x)) = x  and  f(f_inv(x)) = x')
print()
for x in range(-3, 4):
    through_f     = f(x)
    round_trip    = f_inv(through_f)
    matches = abs(round_trip - x) < 1e-10
    print(f'  x={x:2}  f({x:2})={through_f:3}  f_inv({through_f:3})={round_trip:.1f}  restored: {matches}')

print()
print('-' * 55)
print()
print('f(x) = x^2: does NOT have a unique inverse')
print()
def f_sq(x):
    return x**2

for x in [4, -4]:
    print(f'  f({x:3}) = {f_sq(x)}   <- both map to the same output')

print()
print('Restricting domain to x >= 0: f_inv(x) = sqrt(x)')
def f_sq_inv(x):
    return math.sqrt(x)   # only valid on x >= 0

for y in [0, 1, 4, 9, 16]:
    inv = f_sq_inv(y)
    back = f_sq(inv)
    print(f'  f_inv({y:2}) = {inv:.1f}   f(f_inv({y:2})) = {back:.0f}')
````

## Formal Notation

A function $f$ from set $A$ to set $B$ is written $f: A \to B$. For each $x \in A$, $f$ assigns exactly one $f(x) \in B$.

- **Domain**: the set $A$ of valid inputs
- **Codomain**: the set $B$ that outputs are drawn from
- **Range**: the subset of $B$ actually produced by the function (may be smaller than the codomain)
- **$f(x)$**: the output produced by input $x$; also called the **image** of $x$ under $f$

**Composition**: $(f \circ g)(x) = f(g(x))$. Apply $g$ first, then $f$.

**Inverse**: $f^{-1}$ exists (as a function) if and only if $f$ is **one-to-one** (injective) — no two inputs produce the same output. When it exists: $f^{-1}(f(x)) = x$ for all $x$ in the domain.

Note: $f^{-1}(x)$ denotes the inverse function. It does **not** mean $\frac{1}{f(x)}$ — that would be written $(f(x))^{-1}$ or $\frac{1}{f(x)}$.

---

## Real-World Connection

- **Type systems**: a function's type signature is its domain and codomain made explicit. `def f(x: int) -> float` declares the domain is integers, the codomain is floats. A type error is a domain violation.
- **Middleware and pipelines**: every middleware layer in a web framework is a function. The pipeline `auth -> validate -> handler -> serialise` is function composition. The order is not interchangeable.
- **Encryption**: every encryption scheme is a function $f$ (plaintext $\to$ ciphertext) paired with its inverse $f^{-1}$ (ciphertext $\to$ plaintext). Without the inverse being a proper function — uniquely reversible — decryption is impossible.
- **Database indices**: a hash function maps keys to bucket positions. It has a restricted domain (the key type) and a fixed range (bucket indices). Hash collisions are cases where distinct inputs map to the same output — the function is not one-to-one, so no unique inverse exists.
- **Physics and signal processing**: the Fourier transform is a function that maps a signal to its frequency representation. Its inverse — the inverse Fourier transform — undoes this exactly. $f^{-1}(f(x)) = x$ in the most literal sense.

---

## Summary

- A **function** is a rule that maps each input to exactly one output. This is the same rule enforced by pure functions in programming.
- A **relation** is any input-output mapping. A function is a relation where no input appears more than once.
- The **domain** is the set of valid inputs; the **range** is the set of outputs produced. Some functions have restricted domains (division by zero, square root of negative).
- The same function can be expressed as an equation, a table, a graph, or code. These are four views of the same object.
- **Function notation**: $f(x)$ calls function $f$ with input $x$. Substitution replaces every $x$ — so $f(x+1) \neq f(x) + 1$ in general.
- **Composition** $f(g(x))$: apply $g$ first, then $f$. Order matters — $f(g(x)) \neq g(f(x))$ in general.
- **Inverse** $f^{-1}$: the function that undoes $f$. Found by swapping $x$ and $y$ in the equation and solving for $y$. Exists only when $f$ is one-to-one.
