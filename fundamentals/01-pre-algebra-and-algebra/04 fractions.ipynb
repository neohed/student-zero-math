# Adding and Subtracting Fractions (Same Denominator)

Adding fractions with the same denominator is straightforward. The interesting case is when negative fractions are involved — specifically, what happens when you subtract a negative fraction.

This notebook covers same-denominator fraction arithmetic and builds a solid understanding of the additive inverse, which is what makes the sign rules work. Unlike denominators are covered in the next notebook.

**Prerequisites:** Notebooks 01–03. A basic idea of what a fraction represents.

---

## Intuition: A Fraction is a Count of Equal-Sized Parts

Think of a fraction `a/b` as: *`a` chunks, where each chunk is `1/b` of the whole*.

The denominator defines the chunk size. The numerator counts how many chunks you have.

```python
# 3/8 means: 3 chunks, each of size 1/8
chunk_size = 1/8
count      = 3
value      = count * chunk_size  # => 0.375
```

When the denominator is the same, the chunk size is the same — so adding and subtracting is just counting chunks. You would not add centimetres to inches without converting first; same idea.

```python
# 2/8 + 3/8: same chunk size, just add the counts
# => 5 chunks of size 1/8 => 5/8
```

The denominator does not change. Only the count (numerator) changes.

````python
from fractions import Fraction

# Addition: same denominator — add the numerators, keep the denominator
examples_add = [
    (Fraction(2, 8),  Fraction(3, 8)),
    (Fraction(1, 5),  Fraction(3, 5)),
    (Fraction(-3, 7), Fraction(1, 7)),
    (Fraction(-2, 9), Fraction(-4, 9)),
]

print("Addition (same denominator):")
for a, b in examples_add:
    print(f"  {a} + {b} = {a + b}")

# Subtraction: same denominator — subtract the numerators, keep the denominator
examples_sub = [
    (Fraction(5, 8),  Fraction(3, 8)),
    (Fraction(3, 5),  Fraction(4, 5)),
    (Fraction(-1, 6), Fraction(2, 6)),
]

print("\nSubtraction (same denominator):")
for a, b in examples_sub:
    print(f"  {a} - {b} = {a - b}")
````

## The Sign Case: Subtracting a Negative

The rule you already know: **subtracting a negative is the same as adding a positive**.

The reason comes directly from the additive inverse, which we first saw in notebook 02. Recall:

> The **additive inverse** of `x` is `-x`. Adding them gives zero: `x + (-x) == 0`.

When you write `a - (-b)`, you are subtracting `-b`. Subtracting something means adding its additive inverse. The additive inverse of `-b` is `+b`. So:

```
a - (-b)
= a + (additive inverse of -b)
= a + b
```

The double negative cancels because each negation flips the sign once.

````python
# Demonstrate the double-negative cancellation

from fractions import Fraction

cases = [
    (Fraction(-5, 7), Fraction(-1, 7)),   # (-5/7) - (-1/7)
    (Fraction(2, 9),  Fraction(-5, 9)),   # (2/9)  - (-5/9)
    (Fraction(3, 8),  Fraction(-2, 8)),   # (3/8)  - (-2/8)
]

print(f"{'Expression':<25}  {'= (add inverse)':25}  {'Result'}")
print("-" * 65)
for a, b in cases:
    expression    = f"{a} - ({b})"
    rewritten     = f"{a} + ({-b})"
    result        = a - b
    print(f"{expression:<25}  {rewritten:<25}  {result}")
````

The table makes the pattern clear: when the second fraction is negative, subtracting it adds the positive version. When it is positive, subtracting it works as expected.

## Visualisation: Fractions on a Number Line

A number line shows addition and subtraction as movement — right for positive, left for negative. Subtracting a negative means moving right twice, which is the same as moving further right once.

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from fractions import Fraction

def plot_fraction_arithmetic(a: Fraction, op: str, b: Fraction):
    """Visualise a +/- b on a number line, showing both operands as arrows."""
    result = a + b if op == '+' else a - b
    step = b if op == '+' else -b  # actual movement

    fig, ax = plt.subplots(figsize=(10, 2))

    margin = 0.3
    lo = min(0, float(a), float(result)) - margin
    hi = max(0, float(a), float(result)) + margin

    ax.axhline(0.5, color='black', linewidth=1)
    ax.set_xlim(lo, hi)
    ax.set_ylim(0, 1)

    # Tick marks at each fraction unit
    denom = a.denominator
    tick_lo = int(lo * denom) - 1
    tick_hi = int(hi * denom) + 2
    for n in range(tick_lo, tick_hi):
        xpos = n / denom
        ax.plot(xpos, 0.5, '|', color='gray', markersize=8)
        if n % denom == 0 or abs(n) <= denom * 2:
            label = str(Fraction(n, denom))
            ax.text(xpos, 0.3, label, ha='center', va='top', fontsize=8, color='gray')

    # Arrow: start at 0 -> a
    ax.annotate('', xy=(float(a), 0.7), xytext=(0, 0.7),
                arrowprops=dict(arrowstyle='->', color='steelblue', lw=2))
    ax.text(float(a) / 2, 0.82, str(a), ha='center', fontsize=9, color='steelblue')

    # Arrow: a -> result
    ax.annotate('', xy=(float(result), 0.55), xytext=(float(a), 0.55),
                arrowprops=dict(arrowstyle='->', color='darkorange', lw=2))
    mid = (float(a) + float(result)) / 2
    sign_str = f"{'+'  if op == '+' else '-'} ({b})"
    ax.text(mid, 0.62, sign_str, ha='center', fontsize=9, color='darkorange')

    # Mark result
    ax.plot(float(result), 0.5, 'o', color='green', markersize=8, zorder=5)
    ax.text(float(result), 0.15, f'= {result}', ha='center', fontsize=10,
            color='green', fontweight='bold')

    ax.set_yticks([])
    ax.set_title(f'{a} {op} ({b}) = {result}', fontsize=11)
    ax.spines[['top', 'right', 'left']].set_visible(False)
    plt.tight_layout()
    plt.show()

# Straightforward subtraction
plot_fraction_arithmetic(Fraction(5, 7), '-', Fraction(2, 7))

# Subtracting a negative — the arrow goes right, not left
plot_fraction_arithmetic(Fraction(-5, 7), '-', Fraction(-1, 7))

# Adding a negative — same as subtracting the positive
plot_fraction_arithmetic(Fraction(3, 8), '+', Fraction(-5, 8))
````

## Step-by-Step Breakdown

Here is a function that shows each operation in plain steps — useful for tracing the logic when signs get confusing.

````python
from fractions import Fraction

def explain_subtraction(a: Fraction, b: Fraction):
    """Print a step-by-step explanation of a - b."""
    print(f"  {a} - ({b})")
    if b < 0:
        print(f"  = {a} + ({-b})   <- subtracting {b} means adding its inverse, {-b}")
    else:
        print(f"  = {a} + ({-b})   <- subtracting {b} means adding its inverse, {-b}")
    result = a - b
    print(f"  = {result}")
    print()

explain_subtraction(Fraction(-5, 7), Fraction(-1, 7))
explain_subtraction(Fraction(2, 9),  Fraction(-5, 9))
explain_subtraction(Fraction(-3, 4), Fraction(1, 4))
````

## Formal Notation

For fractions with the same denominator $d$:

$$\frac{a}{d} + \frac{b}{d} = \frac{a + b}{d}$$

$$\frac{a}{d} - \frac{b}{d} = \frac{a - b}{d}$$

The sign rule follows from the definition of subtraction:

$$a - (-b) = a + b$$

Because subtraction is defined as adding the additive inverse, and the additive inverse of $-b$ is $+b$:

$$a - (-b) \equiv a + \underbrace{(-1) \times (-b)}_{= +b} = a + b$$

---

## Real-World Connection

- **Financial calculations**: a transaction of $-\frac{1}{4}$ (a debit of a quarter) subtracted from a balance means the debit is reversed — the balance goes up. `balance - (-1/4)` = `balance + 1/4`. The same sign rule.
- **Temperature deltas**: if the temperature change is $-3°$ and you're removing that change (correcting a reading), you add $+3°$. Subtracting a negative change.
- **Signed offsets in graphics**: in coordinate systems, subtracting a negative offset moves an object in the positive direction. Every UI layout engine uses this arithmetic constantly.

## Summary

- When denominators match, add or subtract the numerators — the denominator stays the same
- Subtracting a fraction means adding its additive inverse
- Subtracting a *negative* fraction flips to addition: `a - (-b) = a + b`
- This is not a special rule — it follows directly from the definition of subtraction and the additive inverse
