# Dividing by a Fraction: The Reciprocal Rule

Dividing by a fraction confuses a lot of people. The result seems backwards — dividing by a small number gives a *larger* answer. This notebook builds the intuition for why, then shows the rule that makes the calculation mechanical.

The rule itself follows directly from the reciprocal (notebook 02): dividing by a number is the same as multiplying by its reciprocal. The only new part is applying that idea to fractions.

**Prerequisites:** Notebook 01 (Arithmetic Properties), Notebook 02 (The Reciprocal).

---

## Intuition: Division as Grouping

Division answers the question: **how many groups of this size fit into that total?**

You know this from integer division in Python:

```python
10 // 2   # => 5  (how many groups of 2 fit in 10?)
10 // 5   # => 2  (how many groups of 5 fit in 10?)
```

The same question works for fractions:

```python
1 / 0.5   # => 2.0  (how many halves fit in 1?)
1 / 0.25  # => 4.0  (how many quarters fit in 1?)
```

Dividing by a *smaller* number gives a *larger* answer — because more small groups fit into the total. This is why `6 / (1/3)` gives `18`: there are 18 thirds in 6.

The reciprocal rule is just a clean way to calculate this without thinking in decimals.

````python
# Division as grouping — the pattern holds whether the divisor is a whole number or a fraction

examples = [
    (10,   2,   "how many 2s in 10?"),
    (10,   0.5, "how many halves in 10?"),
    (1,    0.25,"how many quarters in 1?"),
    (6,    1/3, "how many thirds in 6?"),
    (3/4,  1/4, "how many quarters in 3/4?"),
]

for dividend, divisor, question in examples:
    result = dividend / divisor
    print(f"  {dividend} / {divisor:.4g} = {result:.4g}   ({question})")
````

## The Rule: Flip and Multiply

In notebook 02 we established that `a / b = a * (1/b)` — dividing is multiplying by the reciprocal. The same thing applies when the divisor is a fraction:

$$\frac{a}{b} \div \frac{c}{d} = \frac{a}{b} \times \frac{d}{c}$$

Flip the second fraction (take its reciprocal) and multiply. That's the whole rule.

The mechanics:
1. Keep the first fraction as-is
2. Flip the second fraction (swap numerator and denominator)
3. Multiply straight across: top × top, bottom × bottom
4. Simplify

````python
from math import gcd
from fractions import Fraction

def divide_fractions(a, b, c, d):
    """Divide a/b by c/d using the reciprocal rule. Returns a simplified Fraction."""
    # Multiplying a/b by the reciprocal of c/d (which is d/c)
    return Fraction(a, b) * Fraction(d, c)

# (2/3) / (4/5)
result = divide_fractions(2, 3, 4, 5)
print(f"(2/3) / (4/5) = {result}")

# Verify against Python's float division
print(f"Float check:   {(2/3) / (4/5):.6f} == {float(result):.6f}  =>  {abs((2/3)/(4/5) - float(result)) < 1e-10}")
````

---

## Why Does Flipping Work?

Here's the algebraic proof, kept short.

Ask: what number $x$ satisfies this?

$$\frac{c}{d} \times x = \frac{a}{b}$$

To isolate $x$, multiply both sides by the reciprocal of $\frac{c}{d}$:

$$x = \frac{a}{b} \times \frac{d}{c}$$

The reciprocal cancels the divisor because $\frac{c}{d} \times \frac{d}{c} = 1$. Multiplying by 1 removes it.

Let's verify that cancellation directly:

````python
from fractions import Fraction

a, b = Fraction(2, 3), Fraction(4, 5)

x = a / b  # Python's Fraction handles this cleanly

# Verify: b * x should equal a
print(f"a = {a}")
print(f"b = {b}")
print(f"x = a / b = {x}")
print(f"b * x = {b} * {x} = {b * x}   (should equal a = {a})")
print(f"Check: {b * x == a}")
````

## Visualisation: Counting Groups on a Number Line

The "how many groups" framing becomes concrete on a number line. Here's `3/4 / 1/4` — how many quarter-lengths fit into three-quarters?

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from fractions import Fraction

def plot_division(dividend: Fraction, divisor: Fraction):
    result = dividend / divisor
    n_groups = int(result)  # number of whole groups that fit

    fig, ax = plt.subplots(figsize=(10, 2.5))

    # Draw the dividend bar
    ax.add_patch(patches.Rectangle((0, 0.4), float(dividend), 0.4,
                                   linewidth=1.5, edgecolor='steelblue', facecolor='lightsteelblue'))
    ax.text(float(dividend) / 2, 0.6, f'dividend = {dividend}',
            ha='center', va='center', fontsize=10, color='steelblue')

    # Overlay divisor-sized segments with alternating colours
    colours = ['darkorange', 'moccasin']
    x = 0.0
    for i in range(int(result * 10)):  # enough iterations
        seg_end = x + float(divisor)
        if seg_end > float(dividend) + 1e-9:
            break
        ax.add_patch(patches.Rectangle((x, 0.4), float(divisor), 0.4,
                                       linewidth=1, edgecolor='black',
                                       facecolor=colours[i % 2], alpha=0.7))
        ax.text(x + float(divisor) / 2, 0.6, str(i + 1),
                ha='center', va='center', fontsize=9, fontweight='bold')
        x = seg_end

    # Axis labels
    tick_step = float(divisor)
    ticks = [i * tick_step for i in range(int(float(dividend) / tick_step) + 2)
             if i * tick_step <= float(dividend) + 1e-9]
    ax.set_xticks(ticks)
    ax.set_xticklabels([str(Fraction(t).limit_denominator(20)) for t in ticks])
    ax.set_xlim(-0.05, float(dividend) + 0.1)
    ax.set_ylim(0, 1)
    ax.set_yticks([])
    ax.set_title(f'{dividend} / {divisor} = {result}  '
                 f'(how many {divisor}-sized groups fit in {dividend}?)')
    ax.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

plot_division(Fraction(3, 4), Fraction(1, 4))
plot_division(Fraction(2, 3), Fraction(1, 6))
plot_division(Fraction(3, 4), Fraction(1, 2))
````

## More Examples

| Problem | Question | Calculation | Answer |
|---------|----------|-------------|--------|
| $\frac{3}{4} \div \frac{1}{2}$ | How many halves in $\frac{3}{4}$? | $\frac{3}{4} \times 2 = \frac{6}{4} = \frac{3}{2}$ | $1\frac{1}{2}$ |
| $\frac{2}{3} \div \frac{1}{6}$ | How many sixths in $\frac{2}{3}$? | $\frac{2}{3} \times 6 = \frac{12}{3} = 4$ | $4$ |
| $\frac{4}{5} \div \frac{2}{3}$ | How many $\frac{2}{3}$s in $\frac{4}{5}$? | $\frac{4}{5} \times \frac{3}{2} = \frac{12}{10} = \frac{6}{5}$ | $1\frac{1}{5}$ |

Notice that dividing by a fraction smaller than 1 gives an answer greater than the dividend — more small groups fit in.

````python
from fractions import Fraction

problems = [
    (Fraction(3, 4), Fraction(1, 2)),
    (Fraction(2, 3), Fraction(1, 6)),
    (Fraction(4, 5), Fraction(2, 3)),
]

for a, b in problems:
    result = a / b
    recip  = Fraction(b.denominator, b.numerator)
    print(f"  {a} / {b}  =  {a} * {recip}  =  {result}")
````

## Formal Notation

For fractions $\frac{a}{b}$ and $\frac{c}{d}$ where $b, c, d \neq 0$:

$$\frac{a}{b} \div \frac{c}{d} = \frac{a}{b} \times \frac{d}{c} = \frac{ad}{bc}$$

This follows directly from the definition of the reciprocal:

$$\frac{a}{b} \div \frac{c}{d} = \frac{a}{b} \times \left(\frac{c}{d}\right)^{-1} = \frac{a}{b} \times \frac{d}{c}$$

The rule is sometimes written as **KFC**: **K**eep the first, **F**lip the second, **C**hange to multiplication. The mnemonic is cheesy but the underlying reason is the reciprocal.

---

## Real-World Connection

- **Rate and ratio problems**: if a task takes $\frac{3}{4}$ of an hour and you want to know how many tasks fit in a $\frac{9}{4}$-hour window, you divide — `Fraction(9, 4) / Fraction(3, 4) = 3`. This is a direct application.
- **Signal processing and sample rates**: converting between sample rates involves dividing ratios. A buffer of $\frac{1}{44100}$ seconds per sample divided by $\frac{1}{48000}$ seconds per output sample gives the resampling ratio.
- **Dimensional analysis**: in physics and engineering, unit cancellation works exactly like fraction division. Metres per second divided by metres gives per second — the units flip and cancel just like the reciprocal rule.

## Summary

- Division asks "how many groups of this size fit into that total?" — true for whole numbers and fractions alike
- Dividing by a fraction smaller than 1 gives a result *larger* than the dividend — more small groups fit in
- The rule: flip the divisor (take its reciprocal) and multiply — because dividing by a number is always multiplying by its reciprocal
- The reciprocal cancels the divisor: $\frac{c}{d} \times \frac{d}{c} = 1$
- Python's `fractions.Fraction` handles all of this exactly, with no floating-point rounding
